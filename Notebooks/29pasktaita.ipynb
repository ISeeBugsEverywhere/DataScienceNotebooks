```python
nypd = pd.read_csv('../DATA/NYPD.csv', parse_dates={'DATA':['OCCUR_DATE', 'OCCUR_TIME']})
nypd['V'] = nypd['DATA'].dt.hour
nypdII = nypd[['DATA', 'V', 'BORO', 'STATISTICAL_MURDER_FLAG']]
nypd_boro = nypdII.groupby('BORO')
boros = nypd_boro.groups.keys()
fig, axis = plt.subplots()
for boro in boros:
    b = nypd_boro.get_group(boro)
    h = b['V'].value_counts().sort_index(ascending=True)
    axis.plot(h.index.tolist(), h.values.tolist(), label=boro, marker='o')
axis.set_xticks(h.index)
axis.legend(loc='best')
plt.show()
    
    
```

```python
nypd = pd.read_csv('../DATA/NYPD.csv', parse_dates={'DATA':['OCCUR_DATE', 'OCCUR_TIME']})
nypd['V'] = nypd['DATA'].dt.hour
nypdII = nypd[['DATA', 'V', 'BORO', 'STATISTICAL_MURDER_FLAG']]
nypd_boro = nypdII.groupby('BORO')
boros = nypd_boro.groups.keys()
fig, axis = plt.subplots()
for boro in boros:
    b = nypd_boro.get_group(boro).query('STATISTICAL_MURDER_FLAG == True')
    h = b['V'].value_counts().sort_index(ascending=True)
    axis.plot(h.index.tolist(), h.values.tolist(), label=boro, marker='o')
v = np.arange(0,24)
axis.set_xticks(v)
axis.legend(loc='best')
axis.set_title('Mirtini pagal rajoną ir valandas')
plt.show()
```

```python
# Su pie plot atvaizduokite, kiek mirtinų nusišaudymų tenka kiekvienam rajonui.
mb = nypdII.query('STATISTICAL_MURDER_FLAG == True')['BORO'].value_counts()
plt.pie(mb.values, labels=mb.index, autopct='%.2f%%')
plt.show()
```

```python
# Atrinkite 2018 metais vyrus, moteris, uždirbusius daugiau nei bendras 
# vidutinis atlyginimas, surūšiuokite pagal amžiaus grupes ir atvaizduokite
# procentinį pasiskirstymą su pie plot. 
# Vyrus, moteris atskiruose pie plotuose.
```

```python
d18 = pd.read_csv('../DATA/DUS2018.csv')[['lytis','amzius','bdu_spalio']]

v = d18['bdu_spalio'].mean()

fage = d18.query('lytis == "F" and bdu_spalio > @v')['amzius'].value_counts().sort_index(ascending=True)
mage = d18.query('lytis == "M" and bdu_spalio > @v')['amzius'].value_counts().sort_index(ascending=True)

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(8,4.5), gridspec_kw={'width_ratios':[1,1]})
#ax1:
#plotting code:
ax1.set_title('Mot.')
ax1.pie(x=fage.values, labels=fage.index, autopct='%.2f%%')
#ax1.legend(loc=1)
#ax2:
#plotting code:
ax2.set_title('Vyr.')
ax2.pie(x=mage.values, labels=mage.index, autopct='%.2f%%')
#ax2.legend(loc='best')
fig.tight_layout()
plt.show()

```

```python
# Upd.:
#  1) parodykite, kiek % moterų, vyrų nuo visos imties uždirba daugiau, nei vidurkis

viso = len(d18)
print(viso)
motDaugiauVid = sum(fage.values)
kitif = viso - motDaugiauVid

vyrDaugiauVid = sum(mage.values)
kitim = viso - vyrDaugiauVid

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(8,4.5), gridspec_kw={'width_ratios':[1,1]})
#ax1:
#plotting code:
ax1.set_title('Mot.')
ax1.pie(x=[motDaugiauVid, kitif], labels=['Mot > vid.', 'Visi kiti'], autopct='%.2f%%')

#ax1.legend(loc=1)
#ax2:
#plotting code:
ax2.set_title('Vyr.')
ax2.pie(x=[vyrDaugiauVid, kitim], labels=['Vyr > vid.', 'Visi kiti'], autopct='%.2f%%')

#ax2.legend(loc='best')
fig.tight_layout()
plt.show()




```

```python
#  2) atsižvelkite į amžiaus grupes, irgi lyginame su visa imtimi kiekvieną amžiaus grupę

motAgeGroups = fage.index.tolist()+['Visi kiti']
motAgeVals = fage.values.tolist()+[kitif]
vyrAgeGroups = mage.index.tolist()+['Visi kiti']
vyrAgeVals = mage.values.tolist()+[kitif]

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(8,4.5), gridspec_kw={'width_ratios':[1,1]})
#ax1:
#plotting code:
ax1.set_title('Mot')
ax1.pie(x=motAgeVals, labels=motAgeGroups, autopct='%.2f%%')
#ax1.legend(loc=1)
#ax2:
#plotting code:
ax2.set_title('Vyr')
ax2.pie(x=vyrAgeVals, labels=vyrAgeGroups, autopct='%.2f%%')

#ax2.legend(loc='best')
fig.tight_layout()
plt.show()
```
