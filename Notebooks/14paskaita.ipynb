```python
f = [1,2,3,4]
m = [-2,5,1,3]

fig, axis = plt.subplots(2,2)

axis[0,0].plot(f,m)
axis[0,0].plot(f, list(map(lambda x: x+0.25, m)))
axis[1,1].plot(f, list(map(lambda x: x+0.25, m)))

plt.show()
```

```python
# paprstesnis būdas:
f = [1,2,3,4]
m = [-2,5,1,3]


plt.plot(f,m, linestyle=':', lw=3, label='Kreivė A')
plt.plot(f, list(map(lambda x: x+0.75, m)), linewidth=5, marker='s', markersize=12, label='Kreivė K')
plt.plot(f, list(map(lambda x: x-0.75, m)), linewidth=5, label='Kreivė F', color="xkcd:robin's egg blue")
# plt.ylim(2,3)
# plt.xlim(2,5)
plt.grid()
plt.legend(loc='best')
plt.xlabel('Čia X ašies pavadinimas')
plt.ylabel('Y ašies pavadinimas, eurais')
plt.title('Grafiko pavadinimas')
plt.show()
```

```python
fig, axis = plt.subplots()
#plotting code:
axis.plot(f,m, linestyle=':', lw=3, label='A')
axis.plot(f, list(map(lambda x: x+0.75, m)), linewidth=5, marker='s', markersize=12, label='K')

# axis.set_xlim(2,5)
# axis.set_ylim(2,3)
axis.set_ylabel('Y, €')
axis.set_xlabel('x, doleriais')
axis.set_title('Grafiko pavadinimas')

axis.legend(loc=(1.1,0.85))
axis.grid()
#axis.ticklabel_format(style='plain')
# fig.tight_layout()
plt.show()
```
