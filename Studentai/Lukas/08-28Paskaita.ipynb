```python
import numpy as np
import matplotlib.pyplot as plt
from manofunkcijos import SolarAnalyzer
import glob
```

```python
dat_failai = glob.glob('../../DATA/OOP_DATA/*.dat')
print(len(dat_failai))
print(dat_failai)

medziagos = []
for failas in dat_failai:
    m = failas.split('\\')[-1].split('_')[0]
    medziagos.append(m)
medziagos = sorted(list(set(medziagos)))
print(medziagos)

# for i in mpar.keys():
# mpar[i] = []
# print(mpar)


# pce_dict = dict.fromkeys(medziagos, []) # sukuria zodyna kur kiekvienam 'key' priskiria ta pati viena sarasa

pce_dict= dict.fromkeys(medziagos, None) # sukuria zodyna kur prie key priskiria None
for i in pce_dict.keys():   # kiekvienam key sukuriame atskira sarara
    pce_dict[i] = []
print(pce_dict)


jsc_dict = dict.fromkeys(medziagos, None)
for i in jsc_dict.keys():   # kiekvienam key sukuriame atskira sarara
    jsc_dict[i] = []
print(jsc_dict)

Uoc_dict = dict.fromkeys(medziagos, None)
for i in Uoc_dict.keys():   # kiekvienam key sukuriame atskira sarara
    Uoc_dict[i] = []
print(Uoc_dict)

FF_dict = dict.fromkeys(medziagos, None)
for i in FF_dict.keys():   # kiekvienam key sukuriame atskira sarara
    FF_dict[i] = []
print(FF_dict)



for onefile in dat_failai:
    data = SolarAnalyzer(onefile, ';')
    name = data.get_name()
    if name == medziagos[0]:
        pce_dict[medziagos[0]].append(data.get_pce())
        jsc_dict[medziagos[0]].append(data.get_jsc())
        Uoc_dict[medziagos[0]].append(data.get_Uoc())
        FF_dict[medziagos[0]].append(data.get_FF())
        
    elif name == medziagos[1]:
        pce_dict[medziagos[1]].append(data.get_pce())
        jsc_dict[medziagos[1]].append(data.get_jsc())
        Uoc_dict[medziagos[1]].append(data.get_Uoc())
        FF_dict[medziagos[1]].append(data.get_FF())
        
    elif name == medziagos[2]:
        pce_dict[medziagos[2]].append(data.get_pce())
        jsc_dict[medziagos[2]].append(data.get_jsc())
        Uoc_dict[medziagos[2]].append(data.get_Uoc())
        FF_dict[medziagos[2]].append(data.get_FF())
        
    elif name == medziagos[3]:
        pce_dict[medziagos[3]].append(data.get_pce())
        jsc_dict[medziagos[3]].append(data.get_jsc())
        Uoc_dict[medziagos[3]].append(data.get_Uoc())
        FF_dict[medziagos[3]].append(data.get_FF())
        
    elif name == medziagos[4]:
        pce_dict[medziagos[4]].append(data.get_pce())
        jsc_dict[medziagos[4]].append(data.get_jsc())
        Uoc_dict[medziagos[4]].append(data.get_Uoc())
        FF_dict[medziagos[4]].append(data.get_FF())
        
        
print(pce_dict)
print(jsc_dict)
print(Uoc_dict)
print(FF_dict)


print(pce_dict['REF'])
print(pce_dict['V1145'])
print(pce_dict['V1177'])
print(pce_dict['V1178'])
print(pce_dict['V1179'])

print(len(pce_dict['REF']))
print(len(pce_dict['V1145']))
print(len(pce_dict['V1177']))
print(len(pce_dict['V1178']))
print(len(pce_dict['V1179']))


# braizom grafikus
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2, figsize=(8,6))

ax1.boxplot(Uoc_dict.values(), showmeans=True, showfliers=True)
ax1.set_xticklabels(medziagos)
ax1.set_ylabel('Uoc[V]')
ax1.grid()

ax2.boxplot(jsc_dict.values(), showmeans=True, showfliers=True)
ax2.set_xticklabels(medziagos)
ax2.set_ylabel('jsc [mA/cm2]')
ax2.grid()

ax3.boxplot(FF_dict.values(), showmeans=True, showfliers=True)
ax3.set_xticklabels(medziagos)
ax3.set_ylabel('FF [%]')
ax3.grid()

ax4.boxplot(pce_dict.values(), showmeans=True, showfliers=True)
ax4.set_xticklabels(medziagos)
ax4.set_ylabel('PCE [%]')
ax4.grid()

fig.tight_layout()
plt.show()
```
