```python
# Atvaizduokite grafiką, kuriame būtų tik mirtinų susiškaudymų kiekiai per valandą kiekviename rajone.
# Panaudokite plot() grafiką.
# Su pie plot atvaizduokite, kiek mirtinų nusišaudymų tenka kiekvienam rajonui.

nypd = pd.read_csv('../../DATA/NYPD.csv', parse_dates={'DATA':['OCCUR_DATE', 'OCCUR_TIME']})
nypd['V'] = nypd['DATA'].dt.hour
nypd2 = nypd[['DATA', 'V', 'BORO', 'STATISTICAL_MURDER_FLAG']]

nypd_df = nypd2[(nypd2['STATISTICAL_MURDER_FLAG'] == 1)]

nypd_boro_flag = nypd_df.groupby(['BORO', 'STATISTICAL_MURDER_FLAG'])
# nypd_boro_flag.head()

# nypd_boro_flag.get_group(('BRONX', True))
# df_x = df[(df['lytis'] == l) & (df['issilavinimas'] == pak)]
# df.query(expr='issilavinimas == @e')

boros = nypd_boro_flag.groups.keys()
# boros = nypd_df.keys()

valandos = np.arange(0, 24)
fig, axis = plt.subplots()

for boro in boros:
    b = nypd_boro_flag.get_group(boro)
    # b = nypd_df.get_group(boro)
    h = b['V'].value_counts().sort_index(ascending=True)
    
    axis.plot(h.index, h.values, label=boro, marker='o')

# axis.set_xticks(h.index)    
axis.set_xticks(valandos)    
axis.legend()
plt.show()
```

```python
# Atrinkite 2018 metais vyrus, moteris, uždirbusius daugiau nei bendras vidutinis atlyginimas,
# surūšiuokite pagal amžiaus grupes ir atvaizduokite procentinį pasiskirstymą su pie plot.
# Vyrus, moteris atskiruose pie plotuose.

dus18 = '../../DATA/DUS2018.csv'
df18 = pd.read_csv(dus18)

df = df18[['lytis', 'amzius', 'bdu_spalio']]

vid_atl = df['bdu_spalio'].mean()


df_vyr = df[(df['lytis'] == 'M') & (df['bdu_spalio'] > vid_atl)]
df_mot = df[(df['lytis'] == 'F') & (df['bdu_spalio'] > vid_atl)]


df_v = df_vyr[['amzius']]
c = df_v['amzius'].value_counts().sort_index()
amzius = c.index
kiekisvyr = c.values

df_m = df_mot[['amzius']]
c = df_m['amzius'].value_counts().sort_index()
amzius = c.index
kiekismot = c.values

fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.pie(x=kiekisvyr, labels=amzius, autopct='%.2f%%')
ax1.set_title('vyrai')

ax2.pie(x=kiekismot, labels=amzius, autopct='%.2f%%')
ax2.set_title('moterys')

plt.show()


```

```python
# amzius = df['amzius'].value_counts().index.sort_values()

dus18 = '../../DATA/DUS2018.csv'
df18 = pd.read_csv(dus18)

df = df18[['lytis', 'amzius', 'bdu_spalio']]

vid_atl = df['bdu_spalio'].mean()

df['daugiau'] = df['bdu_spalio'] > vid_atl


amzius = df['amzius'].value_counts().index.sort_values()

vyr = []
mot =[]
visas_kiekis = len(df)
for amz in amzius:
    mot_kiekis = len(df[(df['lytis'] == 'F') & (df['daugiau'] == True) & (df['amzius'] == amz)])
    visas_kiekis = visas_kiekis - mot_kiekis 
    mot.append(mot_kiekis)
mot.append(visas_kiekis)
    


visas_kiekis = len(df)
for amz in amzius:
    vyr_kiekis = len(df[(df['lytis'] == 'M') & (df['daugiau'] == True) & (df['amzius'] == amz)])
    visas_kiekis = visas_kiekis - vyr_kiekis
    vyr.append(vyr_kiekis)
vyr.append(visas_kiekis)

vyr_label = []
for amz in amzius:
    vyr_label.append(amz)
vyr_label.append('kiti')

mot_label = []
for amz in amzius:
    mot_label.append(amz)
mot_label.append('kiti')

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 6))

ax1.pie(x=vyr, labels=vyr_label, autopct='%.2f%%')
ax1.set_title('vyrai')

ax2.pie(x=mot, labels=mot_label, autopct='%.2f%%')
ax2.set_title('moterys')

plt.show()
```
