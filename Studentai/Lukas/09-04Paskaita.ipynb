```python
times = np.random.randint(0,24,1000)
# intervalai : 0-6 - naktis, 7-10 - rytas, 11-12 - priešpietė,
# 13-14-pietūs, 15-18-pavakarė, 19-22 - vakaras, 23 - 0 - naktis. 
# Parašykite funkciją, kuri iš times sąrašo verčių 
# sugeneruotų tekstų (rytas, vakaras, popietė ... ) sąrašą. 
# Pritaikykite map(). Suskaičiuokite, kiek kartų pasikartojo atitinkami tekstai,
# atvaizduokite tai su bar arba barh.


def dienos_metas(arr):
    laikai =list(map(lambda x: 'naktis' if 0<=x<=6 or x==23 
                 else 'rytas' if 7<=x<=10 
                 else 'priespiete' if 11<=x<=12
                 else 'pietus' if  13<=x<=14 
                 else 'pavakare' if  15<=x<=18 
                 else 'vakaras', arr))
    return laikai

print(dienos_metas(times))


# set(), list(), dict.fromkeys()
# .count()
laikas = dienos_metas(times)

x_laikas = list(set(laikas))
laikas_dict = dict.fromkeys(x_laikas, 0)
print(laikas_dict)

for i in x_laikas:
    laikas_dict[i]=laikas.count(i)
    
print(laikas_dict)

kiekiai = laikas_dict.values()
print(kiekiai)

plt.bar(x_laikas, kiekiai)
plt.show()


```

```python
# prisiminti .dt. modulį pandos:
# prie nypd pridėkite naują stulpelį V, kur būtų tik įvykio valanda
# tada pritaikykite savo rašytą funkciją su .apply() ir sukurkite naują
# stulpelį su tekstais rytas, vakaras, priešpietė
# tada vizualizuokite, kuriuo paros metu kike buvo susišaudymų (bar arba barh)

def f(x):
    if 0<=x<=6 or x>=23:
        return 'naktis'
    elif 7<=x<=10:
        return 'rytas'
    elif 11<=x<=12:
        return 'priešpietė'
    elif 13<=x<=14:
        return 'pietūs'
    elif 15<=x<=18:
        return 'pavakarė'
    elif 19<=x<=22:
        return 'vakaras'
    
nypd = pd.read_csv('../../DATA/NYPD.csv', parse_dates={'DATA':['OCCUR_DATE', 'OCCUR_TIME']})[['DATA','BORO','STATISTICAL_MURDER_FLAG']]
nypd['V'] = nypd['DATA'].dt.hour
# nypd.head()
nypd['metas'] = nypd['V'].apply(f)
# nypd.head()
# c = nypd[nypd['STATISTICAL_MURDER_FLAG'] == True]['metas'].value_counts()
c = nypd['metas'].value_counts()
metas = c.index
kiekis = c.values

fig, axis = plt.subplots()


B1 = axis.bar(metas, kiekis)
axis.bar_label(B1)
axis.legend(loc='best')


plt.style.use(style='default')
# print(plt.style.available)
#bmh, classic, default, fivethirtyeight, ggplot ...
plt.style.use(style='bmh')

plt.show()

```
