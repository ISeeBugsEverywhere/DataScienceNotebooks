```python
medziagos = []
for failas in dat_failai:
    m = failas.split('\\')[-1].split('_')[0]
    medziagos.append(m)
medziagos = list(set(medziagos))
print(medziagos)
```

```python
pce_dict = dict.fromkeys(medziagos, [])
print(pce_dict)
jsc_dict = dict.fromkeys(medziagos, [])
print(jsc_dict)
Uoc_dict = dict.fromkeys(medziagos, [])
print(Uoc_dict)
FF_dict = dict.fromkeys(medziagos, [])
print(FF_dict)
```

```python
for i in dat_failai:
    data = SolarAnalyzer(i, ';')
    print(data.get_name())
```

```python
for onefile in dat_failai:
    data = SolarAnalyzer(onefile, ';')
    if data.get_name() == 'REF':
        pce_dict['REF'].append(data.get_pce())
        jsc_dict['REF'].append(data.get_jsc())
        Uoc_dict['REF'].append(data.get_Uoc())
        FF_dict['REF'].append(data.get_FF())
        
    elif data.get_name() == 'V1145':
        pce_dict['V1145'].append(data.get_pce())
        jsc_dict['V1145'].append(data.get_jsc())
        Uoc_dict['V1145'].append(data.get_Uoc())
        FF_dict['V1145'].append(data.get_FF())
        
    elif data.get_name() == 'V1177':
        pce_dict['V1177'].append(data.get_pce())
        jsc_dict['V1177'].append(data.get_jsc())
        Uoc_dict['V1177'].append(data.get_Uoc())
        FF_dict['V1177'].append(data.get_FF())
        
    elif data.get_name() == 'V1178':
        pce_dict['V1178'].append(data.get_pce())
        jsc_dict['V1178'].append(data.get_jsc())
        Uoc_dict['V1178'].append(data.get_Uoc())
        FF_dict['V1178'].append(data.get_FF())
        
    elif data.get_name() == 'V1179':
        pce_dict['V1179'].append(data.get_pce())
        jsc_dict['V1179'].append(data.get_jsc())
        Uoc_dict['V1179'].append(data.get_Uoc())
        FF_dict['V1179'].append(data.get_FF())
        
        
print(pce_dict)
print(jsc_dict)
print(Uoc_dict)
print(FF_dict)
```
