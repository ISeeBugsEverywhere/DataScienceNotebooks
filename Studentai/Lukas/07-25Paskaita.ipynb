```python
# A) Duotas sąrašas x=[1,2,3,4,5,6,7,8,9]
# Viename lange, skirtingose ašyse atvaizduokite x2 , x ⋅ 3,
# x ⋅ a, čia a - įvedamas vartotojo.
# Panaudokite skirtingas spalvas, linijų tipus.


x=[1,2,3,4,5,6,7,8,9]
a = 5
y1 = list(map(lambda y: y**2, x))
y2 = list(map(lambda y: y*3, x))
y3 = list(map(lambda y: y*a, x))

fig, axis = plt.subplots(1,3, figsize=[12,4])
axis[0].plot(x, y1, color='green', lw=3, ls='-.')
axis[1].plot(x, y2, color='red', lw=3, ls=':')
axis[2].plot(x, y3, color='blue', lw=4, ls='--')

plt.show()

```

```python
# B) Duoti sąrašai x = [1,2,3,4,5], y1=[2,2,0,0,2], y2 =
# [4,3,2,1,-1], y3 = [2,4,9,16,25], y4 = [-1,1,-1,1,-1]
# Atvaizduokite viename lange x~y1, x~y2,
# x~y3, x~y4 kreives (grafikus). Pirmasis grafikas - linijinis, antrasis - taškinis,
# trečiasis - linija su duomenų taškais, ketvirtasis - brūkšninis.
# Spalvos visų turi būti skirtingos. Grafikai, ašys turi turėti
# pavadinimus.
x = [1,2,3,4,5]
y1=[2,2,0,0,2]
y2 = [4,3,2,1,-1]
y3 = [2,4,9,16,25]
y4 = [-1,1,-1,1,-1]

fig, axis = plt.subplots(2, 2)
axis[0,0].plot(x, y1, ls='solid', lw=3, color='red')
axis[0,1].plot(x, y2, ls='dotted', lw=3, color='green')
axis[1,0].plot(x, y3, ls='solid', lw=3, color='blue', marker='s', markersize=12)
axis[1,1].plot(x, y4, ls='--', lw=3, color='black')

axis[0,0].set_xlabel('X asis')
axis[0,0].set_ylabel('Y asis')
axis[0,1].set_xlabel('X asis')
axis[0,1].set_ylabel('Y asis')
axis[1,0].set_xlabel('X asis')
axis[1,0].set_ylabel('Y asis')
axis[1,1].set_xlabel('X asis')
axis[1,1].set_ylabel('Y asis')

fig.tight_layout()
plt.show()
```

```python
# Sugeneruoti sąrašą x, turintį 101 elementą (nuo 0 iki 100).
# Sukurti antrą sąrašą, kuriame būtų skaičiai, pakelti kvadratu, iš
# pirmojo sąrašo (x2 )
# Sukurti trečiąjį sąrašą, kuriame skaičiai būtų pakelti kvadratu ir
# padauginti iš atsitiktinai sugeneruoto skaičiaus (x2 ⋅ a).
# Sugeneruoti 101-to elementų ilgio sąrašą iš atsitiktinių skaičių.
# Visus šiuos sąrašus atvaizduoti grafike. Grafikas turi turėti
# pavadinimą, pavadintos ašys, pakeisti šriftų dydžiai.

# a = np.arange(1,11)
# b = np.random.randint(1,15,10)
# c = np.random.randint(1,15,10)
# d = np.random.randint(1,15,10)

x = np.arange(0, 101)
y1 = list(map(lambda y: y**2, x))
a = np.random.rand()
y2 = list(map(lambda y: y**2 * a, x))
y3 = np.random.randint(0, 10000, 101)

fig, axis = plt.subplots()
axis.plot(x, y1, color='green', lw=3, ls='-.')
axis.plot(x, y2, color='red', lw=3, ls=':')
axis.plot(x, y3, color='blue', lw=1, ls='--')

plt.show()
```
