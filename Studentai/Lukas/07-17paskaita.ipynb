```python
class SimpleObject():
    def __init__(self, x):
        self.X = x
        
    def get_X(self): # metodas
        return self.X # atributo X pasiekimas metodo viduje
    
    def set_X(self, x):
        self.X = x
        
        
r = SimpleObject(42)
print(r.X)

r.set_X(55)

print(r.get_X())
```

```python
import time
class Tarakonas():
    def __init__(self) -> None:
        print('išsirito tarakonas')
        self.gimtadienis=time.time()
    def __del__(self):
        print('Tarakonas mirė. Jis gyveno:',time.time()-self.gimtadienis)
Rudis=Tarakonas()
time.sleep(0.7)
Juodis=Tarakonas()
time.sleep(0.7)
Pilkis=Tarakonas()
print(Rudis.gimtadienis)
print(Juodis.gimtadienis)
print(Pilkis.gimtadienis)
```

```python
# Parašyti funkciją, kuri nuskaitytų pateiktą DAT failą (DATA aplanke)
# ir grąžintų 6-is sąrašus, kuriuose būtų atitinkamų stulpelių informacija.
# Turite tekstą konvertuoti į float'us
# Turite praleisti tas eilutes
# kur pirmame stuleplyje yra ne skaitinis įrašas

import glob

data_path = 'C:/Users/Lukas/Documents/VCS/DataScienceNotebooks/DATA'

def make_data_lists(data_dir):
    # gaunam data failu sarasa
    data_files = glob.glob(data_dir+'/*.dat')
    
    
    def is_float(string):
        '''funkcija patikrina ar string yra float tipo'''
        try:
            float(string)
            return True
        except ValueError:
            return False
    
    # sukuriame 6 sarasus, nes 6 stulpeliai    
    list1 = []  
    list2 = []  
    list3 = []  
    list4 = []  
    list5 = []  
    list6 = [] 
    
    # is kiekvieno data failo duomenys surasomi i atitinkama stulpeli
    for dataset in data_files:
        try:
            with open(dataset, mode='r', encoding='utf-8') as file:
                turinys = file.readlines()
                
                
                
                for i in turinys:
                    line = i.split()
                    if is_float(line[0]):
                        list1.append(float(line[0]))
                        list2.append(float(line[1]))
                        list3.append(float(line[2]))
                        list4.append(float(line[3]))
                        list5.append(float(line[4]))
                        list6.append(line[5])
        except FileNotFoundError as ex:
            print(ex)
        
            
                    
    return [list1, list2, list3, list4, list5, list6]
    


make_data_lists(data_path)
```
