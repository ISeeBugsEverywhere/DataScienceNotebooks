```python
# „class TxtReader() def“

import numpy as np
import matplotlib.pyplot as plt
from dateutil import parser

# Ankstesnės užduoties klasė (modifikuota):

# Parašyti klasę TxtReader skirtą tekstinio failo iš dviejų stulpelių nuskaitymui.
# Esant poreikiui, inicializacijos metu galima nurodyti stulpelių skirtuką, 
# ar konvertuoti pirmąjį stulpelį į datą (datetime objektą), 
# kiek eilučių praleisti nuo failo pradžios. 
# Jei failas turi daugiau nei du stulpelius - galima nurodyti, kuris stulpelis konkrečiai yra x, o kuris y. 
# Atributai x ir y - privatūs.
# Pridėkite metodus, kurie leistų surasti stulpelio y vidurkį, medianą, maksimumą, minimumą.

class TxtReader():
    def __init__(self,failas,sep=',',header=0,xdata=False,ydata=False,xnr=0,ynr=1):
        self.__xnr=xnr
        self.__ynr=ynr
        self.__x=[]
        self.__y=[]
        print('Skaitomas failas:',failas)
        r=np.loadtxt(failas,delimiter=sep,skiprows=header)
        #!!! Kas čia? Veiksmai dubliuojasi skirtingais įrankiais?
        failas = open(failas,mode='r') # Duomenų failas.
        for i in range(0,header):
            print('skipped:',failas.readline(),end='')
        for eil in failas:
            emas=eil.split(sep)
            # print(emas)
            try:
                if xdata :
                    self.__x.append(parser.parse(emas[self.__xnr]))
                else:
                    self.__x.append(float(emas[self.__xnr]))
                if ydata :
                    self.__y.append(parser.parse(emas[self.__ynr]))
                else:
                    self.__y.append(float(emas[self.__ynr]))
            except ValueError:
                print('Nepavyko atpažinti:',eil)
        failas.close()
        # print(self.__x)     
        # print(self.__y)  
        print()
    def x(self):
        return self.__x
    def y(self):
        return self.__y
    def vidurkis(self):
        return np.average(self.__y)
    def mediana(self):
        return np.median(self.__y)
    def daugiausia(self):
        return np.max(self.__y)
    def mažiausia(self):
        return np.min(self.__y)

ttt=TxtReader('../../DATA/fpc.csv',sep=',',header=1,ydata=False,xnr=0,ynr=1)
# print(ttt.mažiausia())
# print(ttt.vidurkis())
# print(ttt.mediana())
# print(ttt.daugiausia())
print(ttt.x())
print(ttt.y())
```

```python
# ^^^ Būtina įvykdyti kodą aukščiau ^^^

# nuskaityti ir atvaizduoti fpc.csv failo stulpelius grafiškai
o1 = TxtReader('../../DATA/fpc.csv',sep=',',header=1,ynr=1)
o2 = TxtReader('../../DATA/fpc.csv',sep=',',header=1,ynr=2)
o3 = TxtReader('../../DATA/fpc.csv',sep=',',header=1,ynr=3)
o4 = TxtReader('../../DATA/fpc.csv',sep=',',header=1,ynr=4)

fig, axis = plt.subplots(figsize=(8,4.5))
#plotting code:
axis.plot(o1.x(),o1.y())
axis.plot(o2.x(),o2.y()) # axis.scatter
axis.plot(o3.x(),o3.y())
axis.plot(o4.x(),o4.y())
#axis.legend(loc='best')
#axis.ticklabel_format(style='plain')
fig.tight_layout()
plt.show()
```

```python
# ^^^ Būtina įvykdyti kodą aukščiau ^^^

print('„Box and whisker“ grafikas')
# https://en.wikipedia.org/wiki/Box_plot

# tas pats fpc.csv failas.
# Su boxplotais atvaizduokite statistinę informaciją apie 
# stulpelius C2-C5

fig,axis=plt.subplots(figsize=(8,4.5))

# a=np.random.randint(0,1000,20)
# b=np.random.randint(0,1000,20)
# c=np.random.randint(0,1000,20)

axis.boxplot([o1.y(),o2.y(),o3.y(),o4.y()],showmeans=True,showfliers=True)
axis.set_xticklabels(['o1','o2','o3','o4'])

plt.show()
```
