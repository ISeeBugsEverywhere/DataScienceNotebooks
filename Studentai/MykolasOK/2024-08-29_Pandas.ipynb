```python
# Išssilavinimo pakopos yra G1, G2, G3 ir G4. Parašykite ciklą, kuris išspausdintų 
# pakopos kodą ir spalio atlyginų medianą bei vidurkį, atskirai vyrams, atskirai moterims.

pakopa=['G1','G2','G3','G4']

print('Lytis | pakopa | mediana | vidurkis')
for pkp in pakopa:
    m=df14.query('lytis=="F" and issilavinimas==@pkp')
    print('Mot',pkp,round(m['bdu_spalio'].median()),round(m['bdu_spalio'].mean()))

    v=df14.query('lytis=="M" and issilavinimas==@pkp')
    print('Vyr',pkp,round(v['bdu_spalio'].median()),round(v['bdu_spalio'].mean()))
```

```python
# Išssilavinimo pakopos yra G1, G2, G3 ir G4. Parašykite ciklą, kuris išspausdintų 
# pakopos kodą ir spalio atlyginų medianą bei vidurkį, atskirai vyrams, atskirai moterims.

pakopa=['G1','G2','G3','G4']

print('Lytis | pakopa | vidurkis | mediana | skirtumas')
for pkp in pakopa:
    df_f=df14[(df14['lytis']=='F')&(df14['issilavinimas']==pkp)]
    mean=df_f['bdu_spalio'].mean()
    median=df_f['bdu_spalio'].median()
    print('Mot',pkp,round(mean),round(median),round(mean-median))

    df_m=df14[(df14['lytis']=='M')&(df14['issilavinimas']==pkp)]
    mean=df_m['bdu_spalio'].mean()
    median=df_m['bdu_spalio'].median()
    print('Vyr',pkp,round(mean),round(median),round(mean-median))
```

```python
c=df14['lytis'].value_counts()
plt.pie(x=c.values,labels=c.index,autopct='%.1f%%')
plt.show()
# print(c)

e=df14['issilavinimas'].value_counts().sort_index(ascending=True)
plt.pie(x=e.values,labels=e.index,autopct='%.0f%%')
plt.show()
# print(e)

# issilavinimas=sorted(list(set(df14['issilavinimas'])))
```

```python
edus = df14['issilavinimas'].value_counts().sort_index(ascending=True).index.to_list()
# print(edus)
for e in edus:
    d = df14.query('issilavinimas == @e')['bdu_spalio'].mean()
    print(f'{e} {d:.2f}')
```

```python
# Parašykite ciklą, kuris surašytų atlyginimų vidurkius bei medianas
# pagal amžiaus grupes (2014 metams, eurais)
# amžiaus grupes gaukite naudodamiesi value_counts()
# tvarkingesniam atvaizdavimui pasinaudokite f-strings
# išvestyje pridėkite dar ir skirtumą tarp vidurkio ir medianos (vidurkis-mediana)

amžiausGr = df14['amzius'].value_counts().sort_index(ascending=True).index.to_list()
# print(amžiausGr)
for am in amžiausGr:
    du = df14.query('amzius == @am')['bdu_spalio']
    # print(du)
    mean=du.mean()
    medn=du.median()
    diff=mean-medn
    print(f'{am:5} {mean:5.0f} {medn:5.0f} {diff:7.2f}')
```

```python
# value_counts() ir pie plot:
#  kiek procentų buvo vyrų, moterų 2014 metais?
#  kiek procentų buvo iš kiekvienos amžiaus grupės?
#  kiek procentų buvo iš kiekvienos išsilavinimo pakopos?
#  kiek respondentų buvo iš privataus (B) bei viešojo (A) sektoriaus? (stulpelis: nuosavybe)

print('Lytis 2014:')
c=df14['lytis'].value_counts()
plt.pie(x=c.values,labels=c.index,autopct='%.1f%%')
plt.show()

print('Amžius:')
c=df14['amzius'].value_counts()
plt.pie(x=c.values,labels=c.index,autopct='%.0f%%')
plt.show()

print('Išsilavinimas:')
e=df14['issilavinimas'].value_counts().sort_index(ascending=True)
plt.pie(x=e.values,labels=e.index,autopct='%.0f%%')
plt.show()

print('Nuosavybė:')
e=df14['nuosavybe'].value_counts().sort_index(ascending=True)
plt.pie(x=e.values,labels=e.index,autopct='%.0f%%')
plt.show()

print('Įmonės dydis:')
e=df14['im_dydis'].value_counts().sort_index(ascending=True)
plt.pie(x=e.values,labels=e.index,autopct='%.0f%%')
plt.show()
```
