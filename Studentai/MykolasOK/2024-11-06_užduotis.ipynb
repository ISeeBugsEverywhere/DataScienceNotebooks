```python
import glob
import pandas
import matplotlib.pyplot as plt
from pathlib import Path

def f_sąrašas( datapath, obj ):
    f_ext='dat'
    print((f'{datapath}/{obj}_*.{f_ext}'))
    return glob.glob(f'{datapath}/{obj}_*.{f_ext}')

def sutvarkytas_df( from_df ):
    # Atrenka IV erdvės ketvirtį:
    df_IV=from_df[(from_df['U[V]']>0)&(from_df['j[mA/cm^2]']<0)]
    # Atrenka reikalingus stulpelius:
    to_df=df_IV[['U[V]','j[mA/cm^2]']]
    return to_df

for i in range(3):
    datapath=f'../../../gautiDuomenys/DATA/DAY{i}'
    lapas,(ax1,ax2)=plt.subplots(1,2,figsize=[10,5])
    plt.suptitle(f'DAY{i}')

    i=0
    minP_P=[]
    df_list=[]
    fname_list=[]
    for failas in f_sąrašas(datapath,'REF'):
        df=pandas.read_csv(failas,sep=';')
        df.columns=df.columns.str.strip()
        minP_P.append(df['P[mW/cm^2]'].min(skipna=True,numeric_only=True))
        fname_list.append(failas)
        df_list.append(df)
        i+=1
    # Atrenka geriausią NK:
    index_of_best=minP_P.index(min(minP_P))
    df_ref=sutvarkytas_df(df_list[index_of_best])
    ax1.plot(df_ref['U[V]'],-df_ref['j[mA/cm^2]'],label=Path(fname_list[index_of_best]).stem)

    for failas in f_sąrašas(datapath,'V1145'):
        df=pandas.read_csv(failas,sep=';')
        df.columns=df.columns.str.strip()
        df_x=sutvarkytas_df(df)
        ax2.plot(df_x['U[V]'],-df_x['j[mA/cm^2]'],label=Path(failas).stem)

    ax1.legend()
    ax2.legend()
    plt.show()
```
