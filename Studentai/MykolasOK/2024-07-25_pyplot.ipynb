```python
import matplotlib.pyplot as plt
import numpy as np

print('Šeši grafikai viename rinkinyje')

L1=list(np.random.randint(low=1,high=9,size=9))
L2=list(np.random.randint(low=1,high=9,size=9))
L3=list(np.random.randint(low=1,high=9,size=9))
L4=list(np.random.randint(low=1,high=9,size=9))

langas,grafikas=plt.subplots(2,3,figsize=[16,8])

grafikas[0,0].plot(L1,L2,label='L1,L2')
grafikas[0,1].plot(L2,L1,label='L2,L1')
grafikas[0,2].plot(L1,L1,label='L1,L1')
grafikas[1,0].plot(L2,L3,label='L2,L3')
grafikas[1,1].plot(L3,L4,label='L3,L4')
grafikas[1,2].plot(L4,L1,label='L4,L1')

for ax_ax in grafikas:
    for ax in ax_ax:
        ax.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7,9]
y=[2,4,8,6,9,7,3,1]

plt.plot(x,y,lw=2,c='green',label='Skaičiai')
plt.plot(y,x,lw=3,c='orange',label='Sukeista')
plt.plot(x,list(map(lambda y:10-y,y)),lw=3,c='cyan',label='Apversta')
plt.scatter(x,list(map(lambda y:y*y/10,y)),lw=3,c='magenta',label='y*y/10')
plt.suptitle('Kelios kreivės viename grafike')
plt.xlim(0,10)
plt.ylim(0,10)
plt.grid()
plt.legend(loc='best')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Užduotys
# A) Duotas sąrašas x=[1,2,3,4,5,6,7,8,9]
# Viename lange, skirtingose ašyse atvaizduokite
# x^2 , x^3, x^a, čia a - įvedamas vartotojo.
# Panaudokite skirtingas spalvas, linijų tipus.

print('A užduotis')

a=2.1
x=[1,2,3,4,5,6,7,8,9]
y1=list(map(lambda xx:xx**2,x))
y2=list(map(lambda xx:xx**3,x))
y3=list(map(lambda xx:a*xx,x))

langas,grafikas=plt.subplots(1,3,figsize=[8,4])

grafikas[0].plot(x,y1,lw=3,c='cyan',label='y=x^2')
grafikas[1].plot(x,y2,lw=3,c='magenta',label='y=x^3')
grafikas[2].plot(x,y3,lw=3,c='green',label='y=x*a')

grafikas[0].grid()
grafikas[1].grid()
grafikas[2].grid()
grafikas[0].legend(loc='best')
grafikas[1].legend(loc='best')
grafikas[2].legend(loc='best')
plt.show()
```

```python
import matplotlib.pyplot as plt

# B) Duoti sąrašai x=[1,2,3,4,5], y1=[2,2,0,0,2], y2=[4,3,2,1,-1], y3=[2,4,9,16,25], y4=[-1,1,-1,1,-1]
# Atvaizduokite viename lange x~y1, x~y2, x~y3, x~y4 kreives (grafikus).
# Pirmasis grafikas - linijinis, antrasis - taškinis,
# trečiasis - linija su duomenų taškais, ketvirtasis - brūkšninis.
# Spalvos visų turi būti skirtingos. Grafikai, ašys turi turėti pavadinimus.

print('B užduotis')

x= [1,2,3,4,5]
y1=[2,2,0,0,2]
y2=[4,3,2,1,-1]
y3=[2,4,9,16,25]
y4=[-1,1,-1,1,-1]

plt.plot(x,y1,lw=2,c='green',label='y1')
plt.plot(x,y2,lw=3,c='orange',label='y2')
plt.plot(x,y3,lw=2,c='brown',label='y3')
plt.plot(x,y4,lw=3,c='cyan',label='y4')

plt.grid()
plt.legend(loc='best')
plt.show()
```
