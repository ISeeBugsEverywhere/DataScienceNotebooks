```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

def draw_room(ax):
    # Kambario matmenys (cm)
    width, depth, height = 400, 325, 300  # Plotis, gylis, auktis
    
    # Kambario kampai
    vertices = [
        [(0, 0, 0), (width, 0, 0), (width, depth, 0), (0, depth, 0)],  # Grindys
        [(0, 0, height), (width, 0, height), (width, depth, height), (0, depth, height)],  # Lubos
        [(0, 0, 0), (0, 0, height), (0, depth, height), (0, depth, 0)],  # Kair siena
        [(width, 0, 0), (width, 0, height), (width, depth, height), (width, depth, 0)],  # Dein siena
        [(0, depth, 0), (width, depth, 0), (width, depth, height), (0, depth, height)],  # Galin siena
    ]
    
    for verts in vertices:
        ax.add_collection3d(Poly3DCollection([verts], color='gray', alpha=0.3))

def draw_furniture(ax):
    furniture = [
        # Lova (100x200 cm)
        [(50, 50, 0), (150, 50, 0), (150, 250, 0), (50, 250, 0)],
        # Sofa (150x75 cm)
        [(250, 75, 0), (400, 75, 0), (400, 150, 0), (250, 150, 0)],
        # Stalas (100x60 cm)
        [(200, 200, 0), (300, 200, 0), (300, 260, 0), (200, 260, 0)],
        # Spinta (80x50 cm, auktis 200 cm)
        [(320, 10, 0), (400, 10, 0), (400, 60, 0), (320, 60, 0)],
    ]
    
    for verts in furniture:
        ax.add_collection3d(Poly3DCollection([verts], color='brown', alpha=0.7))

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

draw_room(ax)
draw_furniture(ax)

ax.set_xlabel('Plotis (cm)')
ax.set_ylabel('Gylis (cm)')
ax.set_zlabel('Auktis (cm)')
ax.set_xlim([0, 400])
ax.set_ylim([0, 325])
ax.set_zlim([0, 300])
ax.view_init(elev=20, azim=135)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

def plot_2d_layout():
    fig, ax = plt.subplots(figsize=(8, 8))
    
    # Buto perimetras
    walls = [(0, 0), (4, 0), (4, 3.25), (0, 3.25), (0, 0)]
    ax.plot(*zip(*walls), color='black', linewidth=2)
    
    # Durys (100 cm ploio)
    ax.plot([1.5, 2.5], [0, 0], color='brown', linewidth=5, label='Durys')
    
    # Langas (200 cm ploio)
    ax.plot([1, 3], [3.25, 3.25], color='blue', linewidth=5, label='Langas')
    
    # Baldai
    ax.add_patch(plt.Rectangle((0.3, 2.5), 0.8, 0.6, color='gray', alpha=0.5, label='Lova'))
    ax.add_patch(plt.Rectangle((2.5, 2.5), 1, 0.5, color='gray', alpha=0.5, label='Sofa'))
    ax.add_patch(plt.Rectangle((0.3, 0.5), 0.6, 0.5, color='gray', alpha=0.5, label='Spinta'))
    
    # WC ir kriaukl
    ax.add_patch(plt.Circle((3.5, 0.5), 0.3, color='blue', alpha=0.5, label='WC'))
    ax.add_patch(plt.Rectangle((3.2, 1.0), 0.6, 0.3, color='blue', alpha=0.5, label='Kriaukl'))
    
    ax.set_xlim(-0.5, 4.5)
    ax.set_ylim(-0.5, 3.75)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.legend()
    plt.title("Buto 2D planas")
    plt.show()

def plot_3d_visualization():
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(111, projection='3d')
    
    # Kambario sienos
    vertices = [
        [(0, 0, 0), (4, 0, 0), (4, 0, 3), (0, 0, 3)],  # Priekin siena su durimis
        [(0, 3.25, 0), (4, 3.25, 0), (4, 3.25, 3), (0, 3.25, 3)],  # Galin siena su langu
        [(0, 0, 0), (0, 3.25, 0), (0, 3.25, 3), (0, 0, 3)],  # Kair siena
        [(4, 0, 0), (4, 3.25, 0), (4, 3.25, 3), (4, 0, 3)],  # Dein siena
        [(0, 0, 3), (4, 0, 3), (4, 3.25, 3), (0, 3.25, 3)],  # Lubos
    ]
    ax.add_collection3d(Poly3DCollection(vertices, facecolors='lightgray', linewidths=1, edgecolors='k', alpha=0.5))
    
    # Baldai (kubeliai 3D erdvje)
    ax.bar3d(0.3, 2.5, 0, 0.8, 0.6, 0.4, color='gray', alpha=0.7)  # Lova
    ax.bar3d(2.5, 2.5, 0, 1, 0.5, 0.4, color='gray', alpha=0.7)  # Sofa
    ax.bar3d(0.3, 0.5, 0, 0.6, 0.5, 0.9, color='gray', alpha=0.7)  # Spinta
    
    # WC ir kriaukl
    ax.bar3d(3.5, 0.5, 0, 0.3, 0.3, 0.4, color='blue', alpha=0.7)  # WC
    ax.bar3d(3.2, 1.0, 0, 0.6, 0.3, 0.4, color='blue', alpha=0.7)  # Kriaukl
    
    ax.set_xlim(0, 4)
    ax.set_ylim(0, 3.25)
    ax.set_zlim(0, 3)
    ax.set_xlabel("Plotis (m)")
    ax.set_ylabel("Ilgis (m)")
    ax.set_zlabel("Auktis (m)")
    plt.title("Buto 3D vizualizacija")
    plt.show()

plot_2d_layout()
plot_3d_visualization()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

# Buto matmenys (cm)
butas_ilgis: int = 400
butas_plotis: int = 325
butas_aukstis: int = 300

# Dur ir lango pozicijos
lango_ilgis: int = 200  # Langas centre sienos
dur_plotis: int = 90   # Durys centre prieingoje sienoje

# Funkcija brti 2D plan
def bresti_2d_plana() -> None:
    fig, ax = plt.subplots(figsize=(8, 6))
    
    # Sienos
    sienos = [
        [(0, 0), (butas_ilgis, 0)],  # Apaia
        [(butas_ilgis, 0), (butas_ilgis, butas_plotis)],  # Dein
        [(butas_ilgis, butas_plotis), (0, butas_plotis)],  # Virus
        [(0, butas_plotis), (0, 0)]  # Kair
    ]
    
    for siena in sienos:
        ax.plot([siena[0][0], siena[1][0]], [siena[0][1], siena[1][1]], 'black', linewidth=3)
    
    # Langas (priekin siena)
    ax.plot([(butas_ilgis - lango_ilgis) / 2, (butas_ilgis + lango_ilgis) / 2], 
            [0, 0], 'blue', linewidth=5, label='Langas')

    # Durys (galin siena)
    ax.plot([(butas_ilgis - dur_plotis) / 2, (butas_ilgis + dur_plotis) / 2], 
            [butas_plotis, butas_plotis], 'brown', linewidth=5, label='Durys')

    # Vonios kambarys
    wc_ilgis: int = 100
    wc_plotis: int = 150
    wc_x: int = butas_ilgis - wc_ilgis
    wc_y: int = butas_plotis - wc_plotis
    wc = plt.Rectangle((wc_x, wc_y), wc_ilgis, wc_plotis, fill=True, color='lightgray', label="WC")
    ax.add_patch(wc)

    # Kriaukl
    ax.scatter(butas_ilgis - 80, butas_plotis - 20, color='blue', s=100, label="Kriaukl")

    # Sofa
    sofa_x: int = 20
    sofa_y: int = 20
    sofa_ilgis: int = 150
    sofa_plotis: int = 80
    sofa = plt.Rectangle((sofa_x, sofa_y), sofa_ilgis, sofa_plotis, fill=True, color='green', label="Sofa")
    ax.add_patch(sofa)

    # Stalas
    stalas_x: int = 150
    stalas_y: int = 200
    stalas_ilgis: int = 100
    stalas_plotis: int = 60
    stalas = plt.Rectangle((stalas_x, stalas_y), stalas_ilgis, stalas_plotis, fill=True, color='brown', label="Stalas")
    ax.add_patch(stalas)

    # Miegamoji zona (lova)
    lova_x: int = 250
    lova_y: int = 50
    lova_ilgis: int = 140
    lova_plotis: int = 200
    lova = plt.Rectangle((lova_x, lova_y), lova_ilgis, lova_plotis, fill=True, color='red', label="Lova")
    ax.add_patch(lova)

    ax.set_xlim(-20, butas_ilgis + 20)
    ax.set_ylim(-20, butas_plotis + 20)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.legend()
    ax.set_title("Buto 2D planas")
    plt.grid(False)
    plt.show()

# 3D vizualizacija
def bresti_3d_modeli() -> None:
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')

    # Sienos
    sienos_3d = [
        [(0, 0, 0), (butas_ilgis, 0, 0), (butas_ilgis, 0, butas_aukstis), (0, 0, butas_aukstis)],  # Priekin
        [(0, 0, 0), (0, butas_plotis, 0), (0, butas_plotis, butas_aukstis), (0, 0, butas_aukstis)],  # Kair
        [(butas_ilgis, 0, 0), (butas_ilgis, butas_plotis, 0), (butas_ilgis, butas_plotis, butas_aukstis), (butas_ilgis, 0, butas_aukstis)],  # Dein
        [(0, butas_plotis, 0), (butas_ilgis, butas_plotis, 0), (butas_ilgis, butas_plotis, butas_aukstis), (0, butas_plotis, butas_aukstis)]  # Galin
    ]
    
    for siena in sienos_3d:
        ax.add_collection3d(Poly3DCollection([siena], color="lightgray", alpha=0.5))

    # Grindys
    grindys = [(0, 0, 0), (butas_ilgis, 0, 0), (butas_ilgis, butas_plotis, 0), (0, butas_plotis, 0)]
    ax.add_collection3d(Poly3DCollection([grindys], color="gray", alpha=0.3))

    # WC
    wc_3d = [
        (butas_ilgis - 100, butas_plotis - 150, 0),
        (butas_ilgis, butas_plotis - 150, 0),
        (butas_ilgis, butas_plotis, 0),
        (butas_ilgis - 100, butas_plotis, 0)
    ]
    ax.add_collection3d(Poly3DCollection([wc_3d], color="blue", alpha=0.6))

    ax.set_xlim([0, butas_ilgis])
    ax.set_ylim([0, butas_plotis])
    ax.set_zlim([0, butas_aukstis])
    ax.set_xlabel('Ilgis (cm)')
    ax.set_ylabel('Plotis (cm)')
    ax.set_zlabel('Auktis (cm)')
    ax.set_title("Buto 3D modelis")
    
    plt.show()

# Generuojame planus
bresti_2d_plana()
bresti_3d_modeli()
```

```python
def bresti_3d_modeli_pilnai() -> None:
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')

    # Sienos su langais ir durimis
    sienos_3d = [
        [(0, 0, 0), (butas_ilgis, 0, 0), (butas_ilgis, 0, butas_aukstis), (0, 0, butas_aukstis)],  # Priekin
        [(0, 0, 0), (0, butas_plotis, 0), (0, butas_plotis, butas_aukstis), (0, 0, butas_aukstis)],  # Kair
        [(butas_ilgis, 0, 0), (butas_ilgis, butas_plotis, 0), (butas_ilgis, butas_plotis, butas_aukstis), (butas_ilgis, 0, butas_aukstis)],  # Dein
        [(0, butas_plotis, 0), (butas_ilgis, butas_plotis, 0), (butas_ilgis, butas_plotis, butas_aukstis), (0, butas_plotis, butas_aukstis)]  # Galin
    ]
    
    for siena in sienos_3d:
        ax.add_collection3d(Poly3DCollection([siena], color="lightgray", alpha=0.5))

    # Grindys
    grindys = [(0, 0, 0), (butas_ilgis, 0, 0), (butas_ilgis, butas_plotis, 0), (0, butas_plotis, 0)]
    ax.add_collection3d(Poly3DCollection([grindys], color="gray", alpha=0.3))

    # WC zona
    wc_3d = [
        (butas_ilgis - 100, butas_plotis - 150, 0),
        (butas_ilgis, butas_plotis - 150, 0),
        (butas_ilgis, butas_plotis, 0),
        (butas_ilgis - 100, butas_plotis, 0)
    ]
    ax.add_collection3d(Poly3DCollection([wc_3d], color="blue", alpha=0.6))

    # Langas (priekin siena, centras)
    ax.plot([(butas_ilgis - lango_ilgis) / 2, (butas_ilgis + lango_ilgis) / 2], 
            [0, 0], [butas_aukstis / 2, butas_aukstis / 2], 'cyan', linewidth=6, label="Langas")

    # Durys (galin siena, centras)
    ax.plot([(butas_ilgis - dur_plotis) / 2, (butas_ilgis + dur_plotis) / 2], 
            [butas_plotis, butas_plotis], [0, 100], 'brown', linewidth=6, label="Durys")

    # Sofa (staiakampis)
    sofa_corners = [
        (20, 20, 0), (170, 20, 0), (170, 100, 0), (20, 100, 0)
    ]
    ax.add_collection3d(Poly3DCollection([sofa_corners], color="green", alpha=0.7, label="Sofa"))

    # Stalas (staiakampis)
    stalas_corners = [
        (150, 200, 0), (250, 200, 0), (250, 260, 0), (150, 260, 0)
    ]
    ax.add_collection3d(Poly3DCollection([stalas_corners], color="brown", alpha=0.7, label="Stalas"))

    # Lova
    lova_corners = [
        (250, 50, 0), (390, 50, 0), (390, 250, 0), (250, 250, 0)
    ]
    ax.add_collection3d(Poly3DCollection([lova_corners], color="red", alpha=0.7, label="Lova"))

    # Kriaukl (takas)
    ax.scatter(butas_ilgis - 80, butas_plotis - 20, 0, color='blue', s=100, label="Kriaukl")

    ax.set_xlim([0, butas_ilgis])
    ax.set_ylim([0, butas_plotis])
    ax.set_zlim([0, butas_aukstis])
    ax.set_xlabel('Ilgis (cm)')
    ax.set_ylabel('Plotis (cm)')
    ax.set_zlabel('Auktis (cm)')
    ax.set_title("Buto 3D modelis su langais, durimis ir baldais")
    ax.legend()

    plt.show()

# Generuojame atnaujint 3D model
bresti_3d_modeli_pilnai()
```
