```python
print('Kelios kreivs viename grafike')

import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7,9]
y=[2,4,8,6,9,7,3,1]

plt.plot(x,y,lw=2,c='green',label='Skaiiai')
plt.plot(y,x,lw=3,c='orange',label='Sukeista')
plt.plot(x,list(map(lambda y:10-y,y)),lw=3,c='cyan',label='Apversta')
plt.xlim(0,10)
plt.ylim(0,10)
plt.grid()
plt.legend(loc='best')
plt.show()
```
