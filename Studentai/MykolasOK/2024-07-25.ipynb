```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7,9]
y=[2,4,8,6,9,7,3,1]

plt.plot(x,y,lw=2,c='green',label='Skaiiai')
plt.plot(y,x,lw=3,c='orange',label='Sukeista')
plt.plot(x,list(map(lambda y:10-y,y)),lw=3,c='cyan',label='Apversta')
plt.scatter(x,list(map(lambda y:y*y/10,y)),lw=3,c='magenta',label='y*y/10')
plt.suptitle('Kelios kreivs viename grafike')
plt.xlim(0,10)
plt.ylim(0,10)
plt.grid()
plt.legend(loc='best')
plt.show()
```

```python
import matplotlib.pyplot as plt

# U탑duotys
# A) Duotas sra큄as x=[1,2,3,4,5,6,7,8,9]
# Viename lange, skirtingose a큄yse atvaizduokite
# x^2 , x^3, x^a, ia a - 캄vedamas vartotojo.
# Panaudokite skirtingas spalvas, linij킬 tipus.

print('A u탑duotis')

a=2.1
x=[1,2,3,4,5,6,7,8,9]
y1=list(map(lambda xx:xx**2,x))
y2=list(map(lambda xx:xx**3,x))
y3=list(map(lambda xx:a*xx,x))

langas,grafikas=plt.subplots(1,3,figsize=[8,4])

grafikas[0].plot(x,y1,lw=3,c='cyan',label='y=x^2')
grafikas[1].plot(x,y2,lw=3,c='magenta',label='y=x^3')
grafikas[2].plot(x,y3,lw=3,c='green',label='y=x*a')

grafikas[0].grid()
grafikas[1].grid()
grafikas[2].grid()
grafikas[0].legend(loc='best')
grafikas[1].legend(loc='best')
grafikas[2].legend(loc='best')
plt.show()
```

```python
import matplotlib.pyplot as plt

# B) Duoti sra큄ai x=[1,2,3,4,5], y1=[2,2,0,0,2], y2=[4,3,2,1,-1], y3=[2,4,9,16,25], y4=[-1,1,-1,1,-1]
# Atvaizduokite viename lange x~y1, x~y2, x~y3, x~y4 kreives (grafikus).
# Pirmasis grafikas - linijinis, antrasis - ta큄kinis,
# treiasis - linija su duomen킬 ta큄kais, ketvirtasis - br큰k큄ninis.
# Spalvos vis킬 turi b큰ti skirtingos. Grafikai, a큄ys turi turti pavadinimus.

print('B u탑duotis')

x= [1,2,3,4,5]
y1=[2,2,0,0,2]
y2=[4,3,2,1,-1]
y3=[2,4,9,16,25]
y4=[-1,1,-1,1,-1]

plt.plot(x,y1,lw=2,c='green',label='y1')
plt.plot(x,y2,lw=3,c='orange',label='y2')
plt.plot(x,y3,lw=2,c='brown',label='y3')
plt.plot(x,y4,lw=3,c='cyan',label='y4')

plt.grid()
plt.legend(loc='best')
plt.show()
```
