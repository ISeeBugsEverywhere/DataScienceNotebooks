```python
import time

class Tarakonas():
    def __init__(self) -> None:
        print('Išsirito tarakonas.')
        self.gimtadienis=time.time()
    def __del__(self):
        print('Tarakonas mirė. Jis gyveno:',time.time()-self.gimtadienis,'ms')

Rudis=Tarakonas()
time.sleep(0.3)
Juodis=Tarakonas()
time.sleep(0.3)
Pilkis=Tarakonas()
time.sleep(0.3)

print('Rudis:',Rudis.gimtadienis)
print('Juodis:',Juodis.gimtadienis)
print('Pilkis:',Pilkis.gimtadienis)

type(Pilkis)

del Pilkis
# del Rudis
del Juodis

# type(Rudis) 
# Po destruktoriaus: NameError: name 'Rudis' is not defined
```

```python

def rodyk_failą( f_vardas ):
    failas = open(f_vardas, mode='r', encoding='utf-8') # Duomenų failas.
    print('-------------')
    for eil in failas:
        print(eil.rstrip())
    print('-------------')
    failas.close()

def is_float(string):
    try:
        float(string)
        return True
    except ValueError:
        return False

def to_float(string):
    try:
        x=float(string)
        return x
    except ValueError:
        return 0

def skaityk_duomenis( f_vardas ):
    failas = open(f_vardas, mode='r', encoding='utf-8')
    rez=[[],[],[],[],[],[]]
    # rez=[[]*6]
    for eil in failas:
        sąrašas=eil.split()
        if not is_float(sąrašas[0]):
            continue
        rez[0].append(float(sąrašas[0]))
        rez[1].append(float(sąrašas[1]))
        rez[2].append(float(sąrašas[2]))
        rez[3].append(float(sąrašas[3]))
        rez[4].append(float(sąrašas[4]))
        rez[5].append(sąrašas[5])
    failas.close()
    # print(rez[0])
    # print(rez[2])
    # print(rez[5])
    return rez


d_failo_v1='..\\..\\DATA\\V1670_2024-04-24.dat'
d_failo_v2='..\\..\\DATA\\V1671_2024-04-24.dat'
# rodyk_failą(d_failo_v2)
duom1=skaityk_duomenis(d_failo_v1)
duom2=skaityk_duomenis(d_failo_v2)

```
