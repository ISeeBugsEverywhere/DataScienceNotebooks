```python
import pandas as pd
import sqlite3
import time
import re
from selenium import webdriver
from selenium.webdriver.firefox.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException, WebDriverException
import random


df = pd.read_csv('varleFridges.csv')
product_urls = df['Link'].tolist()


try:
    service = Service(r'C:\Users\Batia\Downloads\geckodriver-v0.34.0-win64\geckodriver.exe')
    firefox_options = webdriver.FirefoxOptions()
    firefox_options.headless = False 
    firefox_profile = webdriver.FirefoxProfile()

    
    firefox_profile.set_preference("general.useragent.override", "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:92.0) Gecko/20100101 Firefox/92.0")
    firefox_profile.set_preference("dom.webdriver.enabled", False)
    firefox_profile.set_preference('useAutomationExtension', False)
    firefox_profile.update_preferences()

    driver = webdriver.Firefox(service=service, options=firefox_options)
except WebDriverException as e:
    print(f"Error starting the WebDriver: {e}")
    exit()

def random_sleep():
    time.sleep(random.uniform(1.5, 3.5))

def scrape_product_page(url):
    try:
        driver.get(url)
        random_sleep() 
        WebDriverWait(driver, 15).until(EC.presence_of_element_located((By.CLASS_NAME, "title")))
        print(f"Page loaded successfully: {url}")

        
        try:
            title = driver.find_element(By.CSS_SELECTOR, "h1.title[itemprop='name']").text.strip()
        except NoSuchElementException:
            title = None
            print("Product title not found.")

        
        try:
            price_element = driver.find_element(By.CLASS_NAME, "price-value")
            price = price_element.text.strip()
        except NoSuchElementException:
            price = None
            print("Product price not found.")

        
        try:
            brand_element = driver.find_element(By.CSS_SELECTOR, "li.spec-line[itemprop='brand'] span.right-column")
            brand = brand_element.text.strip()
        except NoSuchElementException:
            brand = None
            print("Product brand not found.")

        
        specifications = {} 

        try:
            spec_tables = driver.find_elements(By.CSS_SELECTOR, "table.remove-table-styles")
            for table in spec_tables:
                rows = table.find_elements(By.TAG_NAME, "tr")
                for row in rows:
                    try:
                        cells = row.find_elements(By.TAG_NAME, "td")
                        if len(cells) >= 2:
                            key = cells[0].text.strip()
                            value = cells[1].text.strip()

                            
                            value = re.sub(r'\s*<br\s*/?>\s*', '; ', value)
                            if key and value:
                                specifications[key] = value
                    except NoSuchElementException:
                        continue

            
            spec_divs = driver.find_elements(By.CSS_SELECTOR, "div.center, div.specifications, div.specs")
            for div in spec_divs:
                try:
                    text = div.get_attribute('innerText').strip()
                    lines = [line.strip() for line in text.split('\n') if line.strip()]
                    i = 0
                    while i < len(lines):
                        if ':' in lines[i]:
                            key, value = lines[i].split(':', 1)
                            key = key.strip()
                            value = value.strip()
                            if key and value:
                                specifications[key] = value
                        i += 1
                except Exception as e:
                    continue

           
            spec_lists = driver.find_elements(By.CSS_SELECTOR, "ul.spec-list, ul.features, ul.details, ul.PRODUCT_SPEC_FILTERS")
            for ul in spec_lists:
                list_items = ul.find_elements(By.CLASS_NAME, "spec-line")
                for li in list_items:
                    try:
                        left_column = li.find_element(By.CLASS_NAME, "left-column").text.strip()
                        right_column = li.find_element(By.CLASS_NAME, "right-column").text.strip()
                        if left_column and right_column:
                            specifications[left_column] = right_column
                    except NoSuchElementException:
                        continue

        except Exception as e:
            print(f"Error extracting specifications: {e}")

        
        for key in ["Stabilizavimas", "Maksimalus skrydžio atstumas", "Maksimalus vaizdo perdavimo diapazonas"]:
            if key not in specifications:
                try:
                    specific_element = driver.find_element(By.XPATH, f"//*[contains(text(), '{key}')]")
                    value = specific_element.find_element(By.XPATH, "following-sibling::*").text.strip()
                    if value:
                        specifications[key] = value
                except NoSuchElementException:
                    continue

        
        try:
            fcc_ce_text = driver.find_element(By.XPATH, "//*[contains(text(), 'FCC:') or contains(text(), 'CE:')]").text.strip()
            if fcc_ce_text:
                specifications['FCC/CE Information'] = fcc_ce_text
        except NoSuchElementException:
            pass

        
        try:
            spec_list_items = driver.find_elements(By.CSS_SELECTOR, "ul li.spec-line")
            for item in spec_list_items:
                try:
                    left_column = item.find_element(By.CLASS_NAME, "left-column").text.strip()
                    right_column = item.find_element(By.CLASS_NAME, "right-column").text.strip()
                    if left_column and right_column:
                        specifications[left_column] = right_column
                except NoSuchElementException:
                    continue
        except NoSuchElementException:
            pass

        
        return {
            "title": title,
            "price": price,
            "brand": brand,
            "specifications": specifications
        }

    except TimeoutException:
        print(f"Page load timed out: {url}")
    except WebDriverException as e:
        print(f"WebDriverException occurred: {e}")

    return None


db_connection = sqlite3.connect('varleDrons.db')
cursor = db_connection.cursor()


cursor.execute('''
CREATE TABLE IF NOT EXISTS Product_FridgesMax (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    title TEXT,
    price TEXT,
    brand TEXT
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS Specifications_FridgesMax (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    product_id INTEGER,
    spec_key TEXT,
    spec_value TEXT,
    FOREIGN KEY (product_id) REFERENCES Product_FridgesMax(id)
)
''')


for url in product_urls:
    product_data = scrape_product_page(url)
    if product_data:
        
        cursor.execute('INSERT INTO Product_FridgesMax (title, price, brand) VALUES (?, ?, ?)', 
                       (product_data['title'], product_data['price'], product_data['brand']))
        product_id = cursor.lastrowid

        
        for key, value in product_data['specifications'].items():
            cursor.execute('INSERT INTO Specifications_FridgesMax (product_id, spec_key, spec_value) VALUES (?, ?, ?)', 
                           (product_id, key, value))

        db_connection.commit()


db_connection.close()
driver.quit()
```

```python
import pandas as pd
import sqlite3
import time
import re
import cloudscraper
from bs4 import BeautifulSoup as BS
import random


df = pd.read_csv('piguFridges.csv')
product_urls = df['Link'].tolist()  


scraper = cloudscraper.create_scraper()

def random_sleep():
    time.sleep(random.uniform(1.5, 3.5))

def scrape_product_page(url):
    try:
        
        response = scraper.get(url)
        if response.status_code != 200:
            print(f"Failed to retrieve page: {url}")
            return None

        
        bs = BS(response.content, 'html.parser')

        
        price = None
        try:
            
            price_box = bs.find('div', {'class': 'c-product__price-box'})
            if price_box:
                price_element = price_box.find('div', {'class': 'c-price'})
                if price_element:
                    price = price_element.get_text(separator="\n").strip()
                else:
                    print("Price element not found ")
            else:
                print("Price box element not found.")
        except Exception as e:
            print(f"Error extracting product price: {e}")

        
        brand = None
        try:
            brand_element = bs.find('div', {'class':'c-product__info'}).find('h1', {'class':'c-product__name'})
            if brand_element:
                brand = brand_element.text.strip()
            else:
                print("Brand element not found.")
        except AttributeError:
            print("Error extracting product brand.")

        
        spec_keys = []
        spec_values = []
        try:
            
            spec = bs.find('div', {'class':'h-row h-pl--8 h-pr--8'}).find_all('tr')
            for i in spec:
                td = i.find_all('td')
                key = td[0].text.strip()
                value = td[1].text.strip()
                spec_keys.append(key)
                spec_values.append(value)
        except AttributeError:
            print("Product specifications not found.")

        specifications = dict(zip(spec_keys, spec_values))

        
        return {
            "price": price,
            "brand": brand,
            "specifications": specifications
        }

    except Exception as e:
        print(f"An error occurred while scraping the page: {e}")

    return None


db_connection = sqlite3.connect('varleDrons.db')  
cursor = db_connection.cursor()


cursor.execute('''
CREATE TABLE IF NOT EXISTS Product_PFridges (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    price TEXT,
    brand TEXT
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS Specifications_PFridges (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    product_id INTEGER,
    spec_key TEXT,
    spec_value TEXT,
    FOREIGN KEY (product_id) REFERENCES Product_PFridges(id)
)
''')


for url in product_urls:
    product_data = scrape_product_page(url)
    if product_data:
        
        cursor.execute('INSERT INTO Product_PFridges (price, brand) VALUES (?, ?)', 
                       (product_data['price'], product_data['brand']))
        product_id = cursor.lastrowid

        
        for key, value in product_data['specifications'].items():
            cursor.execute('INSERT INTO Specifications_PFridges (product_id, spec_key, spec_value) VALUES (?, ?, ?)', 
                           (product_id, key, value))

        db_connection.commit()


db_connection.close()

print("Scraping finished!")
```

```python
import pandas as pd
import sqlite3
import time
import re
import cloudscraper
from bs4 import BeautifulSoup as BS


scraper = cloudscraper.create_scraper()


df = pd.read_csv('piguDrons.csv')
product_urls = df['Link'].tolist()  

def random_sleep():
    time.sleep(random.uniform(1.5, 3.5))

def scrape_brandI(url):
    try:        
        response = scraper.get(url)
        if response.status_code != 200:
            print(f"Failed to retrieve page: {url}")
            return None

        
        bs = BS(response.content, 'html.parser')

        
        brandI = None
        try:
            brandI_element = bs.find('div', {'class':'c-accordion-item__title'})
            if brandI_element:
                brandI = brandI_element.text.replace('Prekės aprašymas:', '').strip()
            else:
                print("Brand information not found.")
        except AttributeError:
            print("Error extracting")
        
        return brandI

    except Exception as e:
        print(f"An error while scraping: {e}")

    return None


db_connection = sqlite3.connect('varleDrons.db')  
cursor = db_connection.cursor()


try:
    cursor.execute("alter table Product_PDrons add column brandI TEXT")
except sqlite3.OperationalError:
    
    print("Column 'brandI' already exists, updating.")


for url in product_urls:
    brandI_data = scrape_brandI(url)
    if brandI_data:
        cursor.execute('update Product_PDrons set brandI = ? where id = (select id from Product_PDrons where id = ?)', (brandI, url))
        db_connection.commit()


db_connection.close()

print("finished!")
```

```python
import pandas as pd
import sqlite3
import time
import re
import cloudscraper
from bs4 import BeautifulSoup as BS
import random


df = pd.read_csv('piguFridges.csv')
product_urls = df['Link'].tolist()  


scraper = cloudscraper.create_scraper()

def random_sleep():
    time.sleep(random.uniform(1.5, 3.5))

def scrape_product_page(url):
    try:
        
        response = scraper.get(url)
        if response.status_code != 200:
            print(f"Failed to retrieve page: {url}")
            return None

        
        bs = BS(response.content, 'html.parser')

        
        price = None
        try:
            price_element = bs.find_all('div', {'class': 'c-product__price-box'})
            if price_element:
                price = price_element[0].get_text(separator="\n").strip()
            else:
                print("Product price not found.")
        except IndexError:
            print("Error extracting product price.")

        
        brand = None
        try:
            brand_element = bs.find('a', {'class': 'c-product__brand'})
            if brand_element:
                brand = brand_element.text.strip()
            else:
                print("Brand element not found.")
        except AttributeError:
            print("Error extracting product brand.")

        
        brandI = None
        try:
            brandI_element = bs.find('div', {'class':'c-accordion-item__title'})
            if brandI_element:
                brandI = brandI_element.text.replace('Prekės aprašymas:', '').strip()
            else:
                print("Brand information not found.")
        except AttributeError:
            print("Error extracting additional brand information.")

        
        spec_keys = []
        spec_values = []
        try:
            
            spec = bs.find('div', {'class':'h-row h-pl--8 h-pr--8'}).find_all('tr')
            for i in spec:
                td = i.find_all('td')
                key = td[0].text.strip()
                value = td[1].text.strip()
                spec_keys.append(key)
                spec_values.append(value)
        except AttributeError:
            print("Product specifications not found.")

        specifications = dict(zip(spec_keys, spec_values))

        
        return {
            "price": price,
            "brand": brand,
            "brandI": brandI,
            "specifications": specifications
        }

    except Exception as e:
        print(f"An error occurred while scraping the page: {e}")

    return None


db_connection = sqlite3.connect('varleDrons.db')  
cursor = db_connection.cursor()


cursor.execute('''
CREATE TABLE IF NOT EXISTS Product_PFridges1 (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    price TEXT,
    brand TEXT,
    brandI TEXT
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS Specifications_PFridges1 (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    product_id INTEGER,
    spec_key TEXT,
    spec_value TEXT,
    FOREIGN KEY (product_id) REFERENCES Product_PFridges1(id)
)
''')


for url in product_urls:
    product_data = scrape_product_page(url)
    if product_data:
        
        cursor.execute('INSERT INTO Product_PFridges1 (price, brand, brandI) VALUES (?, ?, ?)', 
                       (product_data['price'], product_data['brand'], product_data['brandI']))
        product_id = cursor.lastrowid

        
        for key, value in product_data['specifications'].items():
            cursor.execute('INSERT INTO Specifications_PFridges1 (product_id, spec_key, spec_value) VALUES (?, ?, ?)', 
                           (product_id, key, value))

        db_connection.commit()


db_connection.close()

print("Scraping finished!")
```

```python
import pandas as pd
import sqlite3
import time
import random
import cloudscraper
from bs4 import BeautifulSoup as BS


df = pd.read_csv('varleFridges.csv')
product_urls = df['Link'].tolist()


scraper = cloudscraper.create_scraper()

def random_sleep():
    time.sleep(random.uniform(1.5, 3.5))

def scrape_drone_specs(url):
    try:
        response = scraper.get(url)
        if response.status_code != 200:
            print(f"Failed to retrieve page: {url}")
            return None

        soup = BS(response.content, 'html.parser')

        
        try:
            brand_element = soup.find('li', {'class': 'spec-line', 'itemprop': 'brand'})
            brand = brand_element.find('span', {'class': 'right-column', 'itemprop': 'name'}).text.strip() if brand_element else None
        except AttributeError:
            brand = None
            print("Brand not found.")

        
        try:
            price_element = soup.find('span', {'class': 'price'})
            price = price_element.text.strip() if price_element else None
        except AttributeError:
            price = None
            print("Price not found.")

        
        spec_keys = []
        spec_values = []

        
        try:
            filters = soup.find('div', {'class': 'PRODUCT_SPEC_FILTERS'}).find_all('li', {'class': 'spec-line'})
            for spec in filters:
                key = spec.find('span', {'class': 'left-column'}).text.strip()
                value_element = spec.find('span', {'class': 'right-column'})
                value = value_element.text.strip() if value_element else None
                spec_keys.append(key)
                spec_values.append(value)
        except AttributeError:
            print("Specifications (PRODUCT_SPEC_FILTERS) not found.")

        
        try:
            specs = soup.find('div', {'class': 'PRODUCT_SPEC'}).find_all('li', {'class': 'spec-line'})
            for spec in specs:
                try:
                    key = spec.find('span', {'class': 'left-column'}).text.strip()
                    value = spec.find('span', {'class': 'right-column'}).text.strip()
                    spec_keys.append(key)
                    spec_values.append(value)
                except AttributeError:
                    try:
                        key = spec.find('div', {'class': 'left-column'}).text.strip()
                        value = spec.find('div', {'class': 'right-column'}).text.strip()
                        spec_keys.append(key)
                        spec_values.append(value)
                    except AttributeError:
                        pass
        except AttributeError:
            print("Specifications (PRODUCT_SPEC) not found.")

        
        specifications = dict(zip(spec_keys, spec_values))

        return {
            "brand": brand,
            "price": price,
            "specifications": specifications
        }

    except Exception as e:
        print(f"An error occurred while scraping the page: {e}")
        return None


db_connection = sqlite3.connect('varleDrons.db')  
cursor = db_connection.cursor()


cursor.execute('''
CREATE TABLE IF NOT EXISTS Products_varleFG1 (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    brand TEXT,
    price TEXT
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS Specifications_varleFG1 (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    product_id INTEGER,
    spec_key TEXT,
    spec_value TEXT,
    FOREIGN KEY (product_id) REFERENCES Products_varleFG1(id)
)
''')


for url in product_urls:
    random_sleep()
    product_data = scrape_drone_specs(url)
    if product_data:
        
        cursor.execute('INSERT INTO Products_varleFG1 (brand, price) VALUES (?, ?)', 
                       (product_data['brand'], product_data['price']))
        db_connection.commit()
        product_id = cursor.lastrowid

        
        specifications = product_data.get('specifications', {})
        for key, value in specifications.items():
            cursor.execute('INSERT INTO Specifications_varleFG1 (product_id, spec_key, spec_value) VALUES (?, ?, ?)', 
                           (product_id, key, value))
        db_connection.commit()


db_connection.close()

print("Scraping finished! Data saved to SQLite database 'varleDrons.db'.")
```
