```python
import pandas as pd
import sqlite3
import time
import re
from selenium import webdriver
from selenium.webdriver.firefox.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException, WebDriverException
import random


df = pd.read_csv('varleBicycles.csv')
product_urls = df['Link'].tolist()


try:
    service = Service(r'C:\Users\Batia\Downloads\geckodriver-v0.34.0-win64\geckodriver.exe')
    firefox_options = webdriver.FirefoxOptions()
    firefox_options.headless = False 
    firefox_profile = webdriver.FirefoxProfile()

    
    firefox_profile.set_preference("general.useragent.override", "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:92.0) Gecko/20100101 Firefox/92.0")
    firefox_profile.set_preference("dom.webdriver.enabled", False)
    firefox_profile.set_preference('useAutomationExtension', False)
    firefox_profile.update_preferences()

    driver = webdriver.Firefox(service=service, options=firefox_options)
except WebDriverException as e:
    print(f"Error starting the WebDriver: {e}")
    exit()

def random_sleep():
    time.sleep(random.uniform(1.5, 3.5))

def scrape_product_page(url):
    try:
        driver.get(url)
        random_sleep() 
        WebDriverWait(driver, 15).until(EC.presence_of_element_located((By.CLASS_NAME, "title")))
        print(f"Page loaded successfully: {url}")

        
        try:
            title = driver.find_element(By.CSS_SELECTOR, "h1.title[itemprop='name']").text.strip()
        except NoSuchElementException:
            title = None
            print("Product title not found.")

        
        try:
            price_element = driver.find_element(By.CLASS_NAME, "price-value")
            price = price_element.text.strip()
        except NoSuchElementException:
            price = None
            print("Product price not found.")

        
        try:
            brand_element = driver.find_element(By.CSS_SELECTOR, "li.spec-line[itemprop='brand'] span.right-column")
            brand = brand_element.text.strip()
        except NoSuchElementException:
            brand = None
            print("Product brand not found.")

        
        specifications = {} 

        try:
            spec_tables = driver.find_elements(By.CSS_SELECTOR, "table.remove-table-styles")
            for table in spec_tables:
                rows = table.find_elements(By.TAG_NAME, "tr")
                for row in rows:
                    try:
                        cells = row.find_elements(By.TAG_NAME, "td")
                        if len(cells) >= 2:
                            key = cells[0].text.strip()
                            value = cells[1].text.strip()

                            
                            value = re.sub(r'\s*<br\s*/?>\s*', '; ', value)
                            if key and value:
                                specifications[key] = value
                    except NoSuchElementException:
                        continue

            
            spec_divs = driver.find_elements(By.CSS_SELECTOR, "div.center, div.specifications, div.specs")
            for div in spec_divs:
                try:
                    text = div.get_attribute('innerText').strip()
                    lines = [line.strip() for line in text.split('\n') if line.strip()]
                    i = 0
                    while i < len(lines):
                        if ':' in lines[i]:
                            key, value = lines[i].split(':', 1)
                            key = key.strip()
                            value = value.strip()
                            if key and value:
                                specifications[key] = value
                        i += 1
                except Exception as e:
                    continue

           
            spec_lists = driver.find_elements(By.CSS_SELECTOR, "ul.spec-list, ul.features, ul.details, ul.PRODUCT_SPEC_FILTERS")
            for ul in spec_lists:
                list_items = ul.find_elements(By.CLASS_NAME, "spec-line")
                for li in list_items:
                    try:
                        left_column = li.find_element(By.CLASS_NAME, "left-column").text.strip()
                        right_column = li.find_element(By.CLASS_NAME, "right-column").text.strip()
                        if left_column and right_column:
                            specifications[left_column] = right_column
                    except NoSuchElementException:
                        continue

        except Exception as e:
            print(f"Error extracting specifications: {e}")

        
        for key in ["Stabilizavimas", "Maksimalus skrydžio atstumas", "Maksimalus vaizdo perdavimo diapazonas"]:
            if key not in specifications:
                try:
                    specific_element = driver.find_element(By.XPATH, f"//*[contains(text(), '{key}')]")
                    value = specific_element.find_element(By.XPATH, "following-sibling::*").text.strip()
                    if value:
                        specifications[key] = value
                except NoSuchElementException:
                    continue

        
        try:
            fcc_ce_text = driver.find_element(By.XPATH, "//*[contains(text(), 'FCC:') or contains(text(), 'CE:')]").text.strip()
            if fcc_ce_text:
                specifications['FCC/CE Information'] = fcc_ce_text
        except NoSuchElementException:
            pass

        
        try:
            spec_list_items = driver.find_elements(By.CSS_SELECTOR, "ul li.spec-line")
            for item in spec_list_items:
                try:
                    left_column = item.find_element(By.CLASS_NAME, "left-column").text.strip()
                    right_column = item.find_element(By.CLASS_NAME, "right-column").text.strip()
                    if left_column and right_column:
                        specifications[left_column] = right_column
                except NoSuchElementException:
                    continue
        except NoSuchElementException:
            pass

        
        return {
            "title": title,
            "price": price,
            "brand": brand,
            "specifications": specifications
        }

    except TimeoutException:
        print(f"Page load timed out: {url}")
    except WebDriverException as e:
        print(f"WebDriverException occurred: {e}")

    return None


db_connection = sqlite3.connect('varleDrons.db')
cursor = db_connection.cursor()


cursor.execute('''
CREATE TABLE IF NOT EXISTS Product_BicyclesMax (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    title TEXT,
    price TEXT,
    brand TEXT
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS Specifications_BicyclesMax (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    product_id INTEGER,
    spec_key TEXT,
    spec_value TEXT,
    FOREIGN KEY (product_id) REFERENCES Product_BicyclesMax(id)
)
''')


for url in product_urls:
    product_data = scrape_product_page(url)
    if product_data:
        
        cursor.execute('INSERT INTO Product_BicyclesMax (title, price, brand) VALUES (?, ?, ?)', 
                       (product_data['title'], product_data['price'], product_data['brand']))
        product_id = cursor.lastrowid

        
        for key, value in product_data['specifications'].items():
            cursor.execute('INSERT INTO Specifications_BicyclesMax (product_id, spec_key, spec_value) VALUES (?, ?, ?)', 
                           (product_id, key, value))

        db_connection.commit()


db_connection.close()
driver.quit()
```
