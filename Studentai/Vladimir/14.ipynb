```python
# A) Duotas sąrašas x=[1,2,3,4,5,6,7,8,9]
# Viename lange, skirtinguose grafikuose atvaizduokite x2 , x ⋅ 3,
# x ⋅ a, čia a - įvedamas vartotojo.
# Panaudokite skirtingas spalvas, linijų tipus.
# B) Duoti sąrašai x = [1,2,3,4,5], y1=[2,2,0,0,2], y2 =
# [4,3,2,1,-1], y3 = [2,4,9,16,25], y4 = [-1,1,-1,1,-1]
# Atvaizduokite viename lange x~y1, x~y2,
# x~y3, x~y4 kreives (grafikus). Pirmasis grafikas - linijinis, antrasis - taškinis,
# trečiasis - linija su duomenų taškais, ketvirtasis - brūkšninis.
# Spalvos visų turi būti skirtingos. Grafikai, ašys turi turėti
# pavadinimus.
```

```python
x=[1,2,3,4,5,6,7,8,9]
a=int(input('Iveskite skaiciu'))

fig, axis =plt.subplots(2, 2)

# plt.plot(f,m, linestyle=':', lw=3 , label= 'Kreive A')
axis[0, 0].plot(x, list(map(lambda y :y**2, x)), linewidth=2 , marker='o' , markersize=8, label='Kreive B', color='#7e1e9c', ls='-.' )
axis[0, 1].plot(x, list(map(lambda y :y*3, x)), linewidth=2 , marker='o' , markersize=8, label='Kreive C' , color='#f97306', ls=':' )
axis[1, 0].plot(x, list(map(lambda y :y*a, x)), linewidth=2 , marker='o' , markersize=8, label='Kreive D' , color='#001146', ls='--' )
plt.grid()



plt.legend(loc='best')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Grafikas')
plt.show()
```

```python
x = [1,2,3,4,5] 
y1=[2,2,0,0,2]
y2 =[4,3,2,1,-1]
y3 = [2,4,9,16,25]
y4 = [-1,1,-1,1,-1]
fig, axis =plt.subplots(2,2)
axis[0,0].plot(x,y1, color='#380282')
axis[0,1].plot(x,y2, ls=':' , color='#7e1e9c')
axis[1,0].plot(x,y3, ls='-.', color='#8eab12')
axis[1,1].plot(x,y4, ls='--', color='#8c000f')

axis[0,0].set_title('First')
axis[0,0].set_xlabel('x')
axis[0,0].set_ylabel('y1')

axis[0,1].set_title('Second')
axis[0,1].set_xlabel('x')
axis[0,1].set_ylabel('y2')

axis[1,0].set_title('Third')
axis[1,0].set_xlabel('x')
axis[1,0].set_ylabel('y3')

axis[1, 1].set_title('Forth')
axis[1, 1].set_xlabel('x')
axis[1, 1].set_ylabel('y4')
fig.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0,101)

l=x**2
d=list(map(lambda y: y*(np.random.rand()), l))
e=np.random.randint(0, 200, 101)

plt.plot(x , label='1')
plt.plot(l , label='2')
plt.plot(d, label='3')
plt.plot(e, label='4')
#plt.plot(f,m, linestyle=':', lw=3 , label= 'Kreive A')
#plt.plot(f, list(map(lambda x :x+0.75, m)), linewidth=2 , marker='o' , markersize=8, label='Kreive B' , color=)
plt.grid()

plt.legend(loc='best')
plt.xlabel('X asis' , fontsize=10)
plt.ylabel('Ygrikas', fontsize=10)
plt.title('Grafiko Pavadinimas' , fontsize=14)
plt.tight_layout()
plt.show()
```
