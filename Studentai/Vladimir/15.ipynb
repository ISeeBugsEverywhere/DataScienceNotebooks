```python
# BrentOilPrices.csv
# x = datetime iš pirmo stulpelio
# y - tiesiog float'ai iš antro stulpelio
# parser.parse() iš dateutil
# plot()


import numpy as np
import matplotlib.pyplot as plt
from dateutil.parser import parse
file_path = r'C:\Users\Batia\courses\DataScienceNotebooks\DataScienceNotebooks\DATA\BrentOilPrices.csv'
lines = np.genfromtxt(file_path, delimiter=',',dtype=None, skip_header=1)
dates = []
prices = []

for line in lines:
        date = parse(line[0])
        price = float(line[1])
        dates.append(date)
        prices.append(price)
plt.figure()
plt.plot(dates, prices, label='Brent Oil Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Brent Oil Prices')
plt.legend()
plt.grid()

plt.show()


```
