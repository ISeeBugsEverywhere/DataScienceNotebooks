```python
txt = 'Labs, šiandien yra 24 liepos, 7-to mėnesio diena'

def isverst(a):
    if a.isnumeric():
        return 'Skacius'
    if a.isalpha():
        return 'S'
    else:
        return 42

l=list(map(isverst, txt))
print(l)
```

```python
import numpy as np
import datetime

class TxtReader:
    def __init__(self, filepath, delimiter=',', date_format=None, skiprows=0, x_col=0, y_col=1):
        self.filepath = filepath
        self.delimiter = delimiter
        self.date_format = date_format
        self.skiprows = skiprows
        self.x_col = x_col
        self.y_col = y_col
        self._x, self._y = self._read_file()
    
    def _read_file(self):
        with open(self.filepath, 'r') as file:
            lines = file.readlines()[self.skiprows:]
        
        x_data = []
        y_data = []
        
        for line in lines:
            columns = line.strip().split(self.delimiter)
            if len(columns) < max(self.x_col, self.y_col) + 1: # Ar stulpeliu linijoje maziau  negu rejkia stulpeliu
                continue
            
            x = columns[self.x_col]
            y = columns[self.y_col]
            


            if self.date_format:
                try:
                    x = datetime.datetime.strptime(x, self.date_format)
                except ValueError:
                    continue
            

            try:
                y = float(y)
            except ValueError:
                continue
            
            x_data.append(x)
            y_data.append(y)
        
        return x_data, y_data
    
    def mean(self):
        return np.mean(self._y)
    
    def median(self):
        return np.median(self._y)
    
    def max(self):
        return np.max(self._y)
    
    def min(self):
        return np.min(self._y)
    
    def get_x(self):
        return self._x
    
    def get_y(self):
        return self._y
filepath= r'C:\Users\Batia\courses\DataScienceNotebooks\DataScienceNotebooks\DATA\RU.csv'

result = TxtReader(filepath, delimiter=',', date_format= '%Y-%m-%d', skiprows=1, x_col=0, y_col=1)
print(result.mean())
print(result.median())
print(result.get_x())
print(result.get_y())
```
