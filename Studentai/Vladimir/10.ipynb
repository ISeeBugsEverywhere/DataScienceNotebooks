```python
# duota skaičių eilutė.
# rasti joje visus skaičius, kurie yra lygūs savo 
# skaitmenų, pakeltų laipsniu 4, arba 5, sumai.
import numpy as np
r = np.random.randint(0,1000,100).tolist()
r = ''.join([str(i) for i in r])
print(r)


def number(num, power):   #tikrina skaiciu ar jis ira savo sk pk lp suma
    
    num_str = str(num)
    suma = sum(int(digit) ** power for digit in num_str)
    return num == suma


def spnumbers(numbers):  # randa visus skaicius atitinkancius kriteriju number
    
    sk = []
    for num in numbers:
        if number(num, 4) or number(num, 5):
            sk.append(num)
    return sk

sk = spnumbers(r)
print("Special numbers:", sk)
```

```python
# Aprašykite klasę Rectangle. Kreipiantis į klasę, reikia nurodyti
# stačiakampio kraštines a ir b.
# Realizuokite atributą Plotas (S = a ⋅ b) ir metodą, grąžinantį
# Ploto reikšmę
# • Taip pat pridėkite metodą, apskaičiuojantį ir grąžinantį
# perimetrą (P = 2(a + b)).

class Rectangle():
    def __init__(self, a, b):
        self.a=a
        self.b=b
    def plotas(self):
        return self.a*self.b
    def perimetras(self):
        return 2*(self.a + self.b)
    
rect=Rectangle(3,7)
print(rect.plotas())
print(rect.perimetras())
```

```python
class Staciakampis:
    def __init__(self, a, b):
        self.a = a
        self.b = b
        self.susk_parametrus()

    def susk_parametrus(self):
        self.perimetras = 2 * (self.a + self.b)
        self.plotas = self.a * self.b

    def atnaujinti_son(self, a, b):
        self.a = a
        self.b = b
        self.susk_parametrus()

    def _perimetras(self):
        return self.perimetras

    def _plotas(self):
        return self.plotas
    
staciakampis = Staciakampis(5, 3)
print("Perimetras:", staciakampis._perimetras())  
print("Plotas:", staciakampis._plotas())            

staciakampis.atnaujinti_son(7, 4)
print("Atnauj per:", staciakampis._perimetras())  
print("Atnauj plotas:", staciakampis._plotas()) 
```

```python
class Kubas:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
        self.susk_parametrus()

    def susk_parametrus(self):
        self.pavir_plotas = 2 * (self.a * self.b + self.b * self.c + self.c * self.a)
        self.turys = self.a * self.b * self.c

    def atnaujinti_son(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
        self.susk_parametrus()

    def _pavir_plotas(self):
        return self.pavir_plotas

    def _turys(self):
        return self.turys
    
kubas = Kubas(2, 3, 4)
print("Pavisiaus plotas:", kubas._pavir_plotas()) 
print("Turys:", kubas._turys())            

kubas.atnaujinti_son(3, 4, 5)
print("Atnaujintas plotas:", kubas._pavir_plotas())
print("Atnaujintas turys:", kubas._turys())  
```
