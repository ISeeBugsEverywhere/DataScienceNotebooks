```python
import sys
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Pirmiausiai sukurkite sąrašą su visų failų keliais, kuriuos norite įkelti
failu_kelias = ['kelias/iki/pirmo/failo.csv', 'kelias/iki/antro/failo.csv']

# Tuščias DataFrame objektas visiems duomenims saugoti
visi_duomenys = pd.DataFrame()

# Eiti per kiekvieną failo kelią sąraše
for kelias in failu_kelias:
    # Įkelia duomenis iš failo
    df = pd.read_csv(kelias, delimiter=',')  # Čia tarkime, kad visi failai naudoja kablelį kaip skirtuką

    # Jei reikia, pridėkite stulpelį, nurodantį duomenų šaltinį
    df['Source'] = os.path.basename(kelias)  # Prideda failo pavadinimą kaip duomenų šaltinį

    # Sujungia įkeltus duomenis su bendru DataFrame
    visi_duomenys = pd.concat([visi_duomenys, df], ignore_index=True)

# Paprašykite vartotojo nurodyti stulpelių indeksus
stulpelis = int(input('Nurodykite stulpelio 1 indeksa (0-3): '))
stulpelis2 = int(input('Nurodykite stulpelio 2 indeksa (0-3): '))

# Nustato stulpelio pavadinimus pagal nurodytus indeksus
x_stulpelio_pavadinimas = visi_duomenys.columns[stulpelis]
y_stulpelio_pavadinimas = visi_duomenys.columns[stulpelis2]

# Brėžia taškinę diagramą
sns.scatterplot(data=visi_duomenys, x=x_stulpelio_pavadinimas, y=y_stulpelio_pavadinimas, hue='Source')
plt.show()
```

```python
import sys 
import os 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

args = sys.argv

ivestis = ''
failai = []



while ivestis != '+':
    ivestis = input('Iveskite failo kelia (arba + jei daugiau nera) ')
    if ivestis != '+':
            failai.append(ivestis)
    else:
        pass

stulp_skirtukas = input('Irasykite stulpeliu skirtukus: ')

duomenys = []

for i in failai: 
    df = pd.read_csv(i, delimiter=stulp_skirtukas)
    duomenys.append(df)

visos_lent = pd.concat(duomenys, ignore_index=True)

print(df)

stulpelis = int(input('Nurodykite stulpelio 1 indeksa (0-3): '))
stulpelis2 = int(input('Nurodykite stulpelio 2 indeksa (0-3): '))

stulpelio_nr = df.iloc[:, stulpelis]
stulpelio_nr2 = df.iloc[:, stulpelis2]

sns.scatterplot(data=visos_lent, x=stulpelio_nr, y=stulpelio_nr2)
plt.show()
```
