```python
# A) Duotas sąrašas x=[1,2,3,4,5,6,7,8,9]
# Viename lange, skirtinguose grafikuose atvaizduokite x2 , x ⋅ 3,
# x ⋅ a, čia a - įvedamas vartotojo.
# Panaudokite skirtingas spalvas, linijų tipus.

a1 = int(input('Iveskite skaiciu')) #vartotojo ivestas skaicius 
a2 = np.arange(1,11)

# b = np.random.randint(1,15,10)  #sarasas 1
x=[1,2,3,4,5,6,7,8,9] # sarasas 2

x2 = [i**2 for i in x]
x3 = [i*3 for i in x]
xa = [i*a1 for i in x]

fig, axis = plt.subplots(2,2)  #nauja lentele 

axis[0,0].plot(x,x2)
axis[0,1].plot(x,x3)
axis[1,0].plot(x,xa)

axis[0,1].legend(loc='best')
axis[1,1].legend(loc='best')
axis[1,0].legend(loc='best')

plt.show()

```

```python
# B. Duoti sąrašai x = [1,2,3,4,5], y1=[2,2,0,0,2], y2 =
# [4,3,2,1,-1], y3 = [2,4,9,16,25], y4 = [-1,1,-1,1,-1]
# Atvaizduokite viename lange x~y1, x~y2,
# x~y3, x~y4 kreives (grafikus). Pirmasis grafikas - linijinis, antrasis - taškinis,
# trečiasis - linija su duomenų taškais, ketvirtasis - brūkšninis.
# Spalvos visų turi būti skirtingos. Grafikai, ašys turi turėti
# pavadinimus.


x = [1,2,3,4,5] 
y1= [2,2,0,0,2]
y2 =[4,3,2,1,-1]
y3 = [2,4,9,16,25]
y4 = [-1,1,-1,1,-1]


fi, axi = plt.subplots()


axi.plot(x,y2)
axi.scatter(x,y1)
axi.plot(x,y4, marker='s')
axi.plot(x,y3, ls = '-.')

axi.set_ylabel('EUR')
axi.set_xlabel('Dollars')
plt.title('grafiko pavadinimas')

plt.show()
```

```python
# Sugeneruoti sąrašą x, turintį 101 elementą (nuo 0 iki 100).
# Sukurti antrą sąrašą, kuriame būtų skaičiai, pakelti kvadratu, iš
# pirmojo sąrašo (x2 )
# Sukurti trečiąjį sąrašą, kuriame skaičiai būtų pakelti kvadratu ir
# padauginti iš atsitiktinai sugeneruoto skaičiaus (x2 ⋅ a).
# Sugeneruoti 101-to elementų ilgio sąrašą iš atsitiktinių skaičių.
# Visus šiuos sąrašus atvaizduoti grafike. Grafikas turi turėti
# pavadinimą, pavadintos ašys, pakeisti šriftų dydžiai.

x = np.arange(1,102)
x2 = x**2 
x3 = x**2 * np.random.randint(1,5)
randomai = np.random.randint(1,1000,101)

fi, axi = plt.subplots()

axi.plot(x,x2)
axi.plot(x,x3)
axi.plot(x,randomai)


axi.set_ylabel('EUR', fontsize = 20)
axi.set_xlabel('Dollars', fontsize = 8)
plt.title('grafiko pavadinimas', fontsize = 30)

plt.show()

```
