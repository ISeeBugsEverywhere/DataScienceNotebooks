```python
import os   # Importuojame os (fail koregavimui)
import glob  # Importuojame glob bibliotek (fail atidarymui)
import numpy as np # Importuojame numpy bibliotek (skaiiavimams su masyvais)
from dateutil import parser   # Importuojame dateutil bibliotekos parser modul (dat ir laik parsavimui)
import matplotlib.pyplot as plt  # Importuojame matplotlib bibliotek grafik braiymui



class TxtReader():  # Pradedame klas TxtReader
    def __init__(self, f, skirtukas=',', ardata=True, praleisti=1, stu1=1, stu2=2): 
        # Inicializuojame klass konstruktori su keliais numatytais parametrais
        self.skirtukas = skirtukas  # Nustatome skirtuko simbol (numatytasis yra kablelis)
        self.ardata = ardata  # Nustatome ar duomenys yra datos tipo (numatytasis yra True)
        self.praleisti = praleisti  # Nustatome, kiek eilui praleisti (numatytasis yra 1)
        self.__x = stu1-1  # Nustatome pirmos stulpelio indeks (pakeiiamas  nuliniu pagrindu)
        self.__y = stu2-1  # Nustatome antro stulpelio indeks (pakeiiamas  nuliniu pagrindu)
        
        os.chdir(r"C:\Users\User\Desktop\Codes\DataScienceNotebooks\DATA")  # Pakeiiame darbo direktorij
        failas = open(f, mode='r', encoding='utf-8')  # Atidarome fail skaitymo reimu
        turinys = failas.readlines()[self.praleisti:]  # Nuskaitome failo turin ir praleidiame pirmsias eilutes
        failas.close()  # Udarome fail
        
        self.c1, self.c2 = [], []  # Inicializuojame du sraus duomenims saugoti
        for line in turinys:
            fs = line.split(skirtukas)  # Padaliname eilut pagal skirtuk
            self.c1.append(fs[self.__x])  # Pridedame pirmo stulpelio reikmes  c1 sra
            self.c2.append(float(fs[self.__y]))  # Pridedame antro stulpelio reikmes  c2 sra kaip float tipo reikmes
            
        if self.ardata:  # Jei ardata yra True, konvertuojame c1 srao reikmes  dat format
            for i in range(len(self.c1)):
                self.c1[i] = parser.parse(self.c1[i]).date()  # Naudojame parser modul dat konvertavimui
                

    def spausdinti_sarasu(self):  # Metodas sra spausdinimui
            print("c1:", self.c1)
            print("c2:", self.c2)
    
a = TxtReader('fpc.csv')  # Sukuriame TxtReader klass objekt ir nurodome fail 'BrentOilPrices.csv'


print(a.spausdinti_sarasu())

```

```python
#textreader class kodas 

from dateutil import parser

class TxtReader():
    def __init__(self, fname, sep=',', use_xy=[0, 1], skip=0, convert_date=False) -> None:
        self.fname = fname
        self.skip = skip
        self.xy = use_xy
        self.sep = sep
        self.convert = convert_date
        self.__X = None
        self.__Y = None
        self.read_file()

    def read_file(self):
        f = open(self.fname, mode='r', encoding='utf-8')
        lines = f.readlines()
        f.close()
        xl = []
        yl = []
        for line in lines[self.skip:]:
            r = line.split(self.sep)
            xi = r[self.xy[0]]
            yi = r[self.xy[1]]
            if self.convert:
                x = parser.parse(xi)
                xl.append(x)
            else:
                x = float(xi)
                xl.append(x)
            y = float(yi)
            yl.append(y)
        self.__X = xl
        self.__Y = yl

    def X(self):
        return self.__X

    def Y(self):
        return self.__Y

# Sukuriame TxtReader klass objekt ir nurodome fail 'fpc.csv'
reader = TxtReader('fpc.csv', sep=',', use_xy=[0, 1], skip=1, convert_date=True)

# Spausdiname X ir Y sraus
print("X (dates or floats):", reader.X())
print("Y (floats):", reader.Y())
    

```
