```python
import sqlite3

conn = sqlite3.connect('AutopliusDuombaze5.db')
c = conn.cursor()
c.execute('''
    CREATE TABLE IF NOT EXISTS Autoplius (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        pavadinimas TEXT,
        kaina INTEGER,
        rida INTEGER,
        amzius TEXT,
        galia INTEGER,
        keb_tipas TEXT,
        pavaru_deze TEXT,
        miestas TEXT,
        salis TEXT,
        ta_galiojimas TEXT,
        kuras TEXT,
        defektai TEXT,
        gamintojas TEXT
    )
''')

conn.commit()
print('lentele sukurta')

# Pridedame duomenis į lentelę iš kiekvieno žodyno sąraše
for z in visi_zodynai:
    c.execute('''
        INSERT INTO Autoplius (pavadinimas, kaina, rida, amzius, galia, keb_tipas, pavaru_deze, miestas, salis, ta_galiojimas, kuras, defektai, gamintojas) 
        VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
    ''', (
        z['pavadinimas'], 
        int(z['kaina']) if z['kaina'] != 'None' else 0,  
        int(z['rida']) if z['rida'] != 'None' else 0,    
        z['amzius'] if z['amzius'] != 'None' else 'Nėra', 
        int(z['galia']) if z['galia'] != 'None' else 0, 
        z['keb_tipas'], 
        z['pavaru_deze'], 
        z['miestas'], 
        z['salis'], 
        z['ta_galiojimas'], 
        z['kuras'], 
        z['defektai'], 
        z['gamintojas']
    ))

# Patvirtiname pakeitimus
conn.commit()

# Uždarome ryšį su DB
conn.close()

print("Duomenys sėkmingai pridėti.")
```
