```python
from bs4 import BeautifulSoup

for i in sk:

    zodynas = {
        'pavadinimas':'None', 'kaina':'None','rida': 'None', 'amzius': 'None', 'galia': 'None', 'keb_tipas': 'None', 
        'pavaru_deze': 'None', 'miestas': 'None', 'salis': 'None', 'ta_galiojimas': 'None', 
        'kuras': 'None', 'defektai': 'None', 'gamintojas': 'None'
    }

    # Pavadinimas ir gamintojas
    sk2 = bs.find('div', {'class':'col-10 page-title'})
    pilnas = sk2.text.strip()
    pavadinimas = pilnas.split(',')[0]
    gamintojas = pilnas.split()[0]
    zodynas['pavadinimas'] = pavadinimas
    zodynas['gamintojas'] = gamintojas

    # Kaina
    sk4 = bs.find('div', {'class':'price'})
    a = int((sk4.text.strip())[:10].replace(' ',''))
    zodynas['kaina'] = a

    # Miestas
    sk5 = bs.find('div', {'class':'owner-location'})
    abc = (sk5.text.strip())
    zodynas['miestas'] = abc


    sk0 = bs.find_all('div', {'class': 'parameter-row'})
    
    for i in sk0:
        label = i.find('div', {'class': 'parameter-label'})
        value = i.find('div', {'class': 'parameter-value'})
    
        if label and value:
            text = label.text.strip()
            v = value.text.strip()
            
            if 'Rida' in text:
                p = int(v.replace(' ', '').replace('km', ''))
                zodynas['rida'] = p

            if 'Pirma registracija' in text:
                q = int(v[:4])
                zodynas['amzius'] = q

            if 'Variklis' in text:
                w = v.split()[0]
                zodynas['galia'] = w

            if 'Kėbulo tipas' in text:
                zodynas['keb_tipas'] = v

            if 'Pavarų dėžė' in text:
                zodynas['pavaru_deze'] = v

            if 'Pirmosios registracijos šalis' in text:
                zodynas['salis'] = v

            if 'Tech. apžiūra iki' in text:
                zodynas['ta_galiojimas'] = v

            if 'Kuro tipas' in text:
                zodynas['kuras'] = v

            if 'Defektai' in text:
                zodynas['defektai'] = v


visi_zodynai.append(zodynas)
```
