```python
# Randame visus 'parameter-row' div elementus
sk = bs.find_all('div', {'class': 'parameter-row'})

# Iteruojame per kiekvieną 'parameter-row' div
for row in sk:
    label = row.find('div', {'class': 'parameter-label'})
    
    # Tikriname, ar elementas su label tekstu egzistuoja ir yra "Pavarų dėžė"
    if label and 'Pavarų dėžė' in label.text:
        # Randame kaimyninį div su klase 'parameter-value'
        value = row.find('div', {'class': 'parameter-value'})
        if value:
            print(value.text.strip())  # Atspausdiname reikšmę (tikimės "Mechaninė")
        break
```
