```python
import sqlite3
import pandas as pd

kaina = []
miestas = []

SDB = sqlite3.connect('WEBscr.db')

sql = "SELECT * FROM TUrl;"
df = pd.read_sql_query(sql, con=SDB)
df2 = df['params'].str.split(';', expand=True)
SDB.close()

# def duomenys(i):
#     a = len(i.dropna())
#     kaina.append(i.iloc[a-1])
#     miestas.append(i.iloc[a-2])

# df['Data_Count'] = df.apply(duomenys, axis=1)

df.head()

for x in df['params']:
    a = x.split(';')
    kaina1 = (a[-1]).replace(' €',"").replace(" ", "")
    miestas1 = a[-2]
    kaina.append(kaina1)
    miestas.append(miestas1)



```

```python
print(miestas)
```

```python
int_kaina = []

for i in kaina: 
    if '\n\n\n\n\n' in i:
        c = i.split('\n\n\n\n\n')
        int_kaina.append((int(c[1])))
        
    else:
        int_kaina.append((int(i)))
```

```python
int_kaina.clear()
```

```python
# CSV atlyginimai

csv_failo_kelias = 'C:\\Users\\User\\Desktop\\Codes\\DataScienceNotebooks\\Studentai\\PauliusS\\Paskaita_52\\atlyginimai.csv'
duomenys = pd.read_csv(csv_failo_kelias, sep =',')

duomenys.head()
```

```python
miest = duomenys.iloc[1:,0].tolist()
atlyg = duomenys.iloc[1:,1].tolist()
print(miest)
print(atlyg)
```

```python
atlyginimai_bruto = []

for i in range(len(atlyg)):
    atlyg[i] = atlyg[i].replace('\xa0', '').replace(',','.')
    v = atlyg[i]
    atlyginimai_bruto.append(float(v))
```

```python
atlyginimai_bruto.clear()
```

```python
print(atlyginimai_bruto)
```

```python
pavadinimai = []

for i in range(len(miest)):
    miest[i] = miest[i].replace('\xa0', '')
    v = miest[i]
    pavadinimai.append(v)
```

```python
print(pavadinimai)
```

```python
miestai2 = list(map(lambda x: x[:4], miestas))
print(miestai2)   #auto miestai 
print(int_kaina)   #auto kaina 
```

```python
pavadinimai2 = list(map(lambda x: x[:4], pavadinimai))
print(pavadinimai2)
print(atlyginimai_bruto)
```

```python
auto_lentele = pd.DataFrame({
    "Miestas": miestai2,
    "Automobiliu_kaina": int_kaina,
})

auto_lentele = auto_lentele.groupby("Miestas", as_index = False).mean()

atlyg_lentele = pd.DataFrame({
    "Miestas" : pavadinimai2,
    "Atlyginimai" : atlyginimai_bruto
       
})

galutine_lentele = pd.merge(atlyg_lentele,auto_lentele, on="Miestas", how="left")
galutine_lentele.head()

plt.figure(figsize=(10,6))
sns.regplot(data=galutine_lentele, x="Atlyginimai", y="Automobiliu_kaina", order = 1)
plt.xlabel("Atlyginimai")
plt.ylabel("Automobiliu kaina")
plt.show()
```
