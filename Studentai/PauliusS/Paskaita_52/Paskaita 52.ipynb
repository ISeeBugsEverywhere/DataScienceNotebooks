```python
import sqlite3
import pandas as pd

kaina = []
miestas = []

SDB = sqlite3.connect('WEBscr.db')

sql = "SELECT * FROM TUrl;"
df = pd.read_sql_query(sql, con=SDB)
df2 = df['params'].str.split(';', expand=True)
SDB.close()

# def duomenys(i):
#     a = len(i.dropna())
#     kaina.append(i.iloc[a-1])
#     miestas.append(i.iloc[a-2])

# df['Data_Count'] = df.apply(duomenys, axis=1)

df.head()

for x in df['params']:
    a = x.split(';')
    kaina1 = (a[-1]).replace(' €',"").replace(" ", "")
    miestas1 = a[-2]
    kaina.append(kaina1)
    miestas.append(miestas1)


int_kaina = []

for i in kaina: 
    if '\n\n\n\n\n' in i:
        c = i.split('\n\n\n\n\n')
        int_kaina.append((int(c[1])))
        
    else:
        int_kaina.append((int(i)))
```

```python
# Atvaizduokite top3 gamintojų skelbimų ir ridų kainų pasiskirstymą su histograma. 
# Panaudokite hue raktažodį bei multipla raktažodį, išbandykite multiple raktažodžio galimas reikšmes. 
# Kuris histogramos variantas yra infromatyviausias?

SDB = sqlite3.connect('AutopliusDuombaze10.db')

sql = "SELECT gamintojas, kaina, rida FROM Autoplius;"
df3 = pd.read_sql_query(sql, con=SDB)
SDB.close()

df3.head()


top_gamintojai = df3['gamintojas'].value_counts().head(4)
top4 = top_gamintojai.index.tolist()

top3 = top4[:2]+top4[3:4]
print(top3)

top4_dataframe = df3[df3['gamintojas'].isin(top3)]
top4_dataframe.head()

sns.histplot(data=top4_dataframe, x='kaina', hue='gamintojas', binwidth=2000, multiple='stack')

plt.xlim(0, 40000)   #apribojame X asi 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sukurkite duomenų rinkinį
data = {'Gamintojas': ['Toyota', 'Honda', 'Nissan', 'BMW'],
        'Pardavimai': [250, 300, 150, 200]}
df = pd.DataFrame(data)

# Sukurkite bar diagramą
ax = sns.barplot(x='Gamintojas', y='Pardavimai', data=df)

# Pridėkite etiketes ant kiekvieno stulpelio
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', color='black', xytext=(0, 5),
                textcoords='offset points')

# Parodykite grafiką
plt.show()
```
