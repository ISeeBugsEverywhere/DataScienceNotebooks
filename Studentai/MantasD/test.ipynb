```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
```

```python
kainos = np.array([600, 5400, 4000, 2000, 2000, 1200])
kategorijos = np.array(['Technologij parinkimas', 'Backend krimas', 'Frontend krimas', 'DevOps ir Diegimas', 'Testavimas', 'Optimizavimas ir SEO'])
data = pd.DataFrame({'Kaina': kainos, 'Kategorija': kategorijos})
data = data.sort_values('Kaina', ascending=False)
plt.figure(figsize=(10, 6))
plt.pie(data['Kaina'], autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Kain pasiskirstymas')
labels = [f'{kategorija} ({kaina} )' for kategorija, kaina in zip(data['Kategorija'], data['Kaina'])]
plt.pie(data['Kaina'], labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Kain pasiskirstymas')
# plt.legend(labels, loc="best", bbox_to_anchor=(1, 1))
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.pie(data['Kaina'], labels=data['Kategorija'], labeldistance=1.1, autopct='%1.1f%%', startangle=140)
plt.title('Kain pasiskirstymas')
plt.show()
```
