```python
s_of_dfs = [iv2013,iv2014,iv2015,iv2016,iv2017,iv2018,iv2019,iv2020,iv2021,iv2022,iv2023]
```

```python
for df in s_of_dfs:
    df['dataLaikas'] = pd.to_datetime(df['registravimoData'])
```

```python
for df in s_of_dfs:
    df['Metai'] = df['dataLaikas'].dt.year
```

```python
def lytis(x):
    for i in x:
        if i['kaltininkas'] == 'Taip':
            return i['lytis']
```

```python
for df in s_of_dfs:
    df['lytis'] = df['eismoDalyviai'].apply(lytis)
```

```python
def nebl(x):
    if x == 'Taip':
        return 1
    else:
        return 0
```

```python
for df in s_of_dfs:
    df['nebalaivus'] = df['neblaivusKaltininkai'].apply(nebl)
```

```python
for df in s_of_dfs:
    df['apsvaige'] = df['apsvaigeKaltininkai'].apply(nebl)
```

```python
# Suraskite 3 populiariausias eismo įvykių schemas! 2023 metais (stulpelis schema2)
iv2023_schemos = iv2023[['schema2', 'Metai']].groupby('schema2').count().reset_index().sort_values(by= 'Metai',ascending=False)[:3]
iv2023_schemos.head()
fig, axes = plt.subplots(figsize=(10,6))
ax = sns.barplot(data=iv2023_schemos,x = 'schema2', y='Metai', ax=axes)
axes.set_title('3 populiariausios schemos')

for container in ax.containers:
    ax.bar_label(container)
```

```python
schemos = list(iv2023_schemos['schema2'])
print(schemos)
```

```python
def schem(x):
    if x in schemos:
        return True
    else:
        return False
```

```python
for df in s_of_dfs:
    df['popschemos'] = df['schema2'].apply(schem)
```

```python
newdf = pd.concat(s_of_dfs) # concat veikia kaip union
```

```python
fig, axes = plt.subplots(figsize=(10,6))

nedf = newdf[newdf['popschemos']== True]
nef = nedf.groupby(['schema2','Metai'])['dataLaikas'].count().reset_index()
axes.set_title('3 populiariausios schemos')
ax = sns.barplot(data=nef,x = 'Metai', y='dataLaikas', ax=axes, hue = 'schema2')
axes.set(xlabel='Metai',ylabel='Kiekiai')
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

```python
fig, axes = plt.subplots(figsize=(10,6))
dfvyrmot = newdf.groupby(['lytis','Metai'])['dataLaikas'].count().reset_index()
axes.set_title('Vyrai-moterys')
ax = sns.barplot(data=dfvyrmot,x = 'Metai', y='dataLaikas', ax=axes, hue = 'lytis')
axes.set(xlabel='Metai',ylabel='Kiekiai')
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

```python
fig, axes = plt.subplots(figsize=(10,6))
dfnebl = newdf[newdf['neblaivusKaltininkai']== 'Taip']
dfvyrmotnebl = dfnebl.groupby(['lytis','Metai'])['dataLaikas'].count().reset_index()
axes.set_title('Vyrai-moterys neblaivūs')
ax = sns.barplot(data=dfvyrmotnebl,x = 'Metai', y='dataLaikas', ax=axes, hue = 'lytis')
axes.set(xlabel='Metai',ylabel='Kiekis')
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```
