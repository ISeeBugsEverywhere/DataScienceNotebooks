```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
import sqlite3
import mysql.connector as cnt
import plotly.express as px
import random

from Skriptai.manofunkcijos import *

import requests
from bs4 import BeautifulSoup
import selenium
from selenium import webdriver # naršyklės kontroleris
from selenium.webdriver.chrome.options import Options #Naršyklės
# pip install beautifulsoup4 selenium
import time
import nordvpn_switcher as ns
```

```python
list_of_urls = ['https://www.varle.lt/saldytuvai/','https://www.varle.lt/dronai/','https://www.varle.lt/plansetiniai-kompiuteriai/','https://www.varle.lt/televizoriai/','https://www.varle.lt/dviraciai/']
opcijos = Options()
opcijos.add_argument('--incognito')
driver = webdriver.Chrome(options=opcijos)
for url in list_of_urls:
    try:
        urlas = url
        driver.get(urlas)
        time.sleep(3)
        source = driver.page_source
        bs = BeautifulSoup(source, 'html.parser')
        try:
            count = int(round(int(bs.find('span', {'class':'results_total_items'}).text.strip().replace('(','').replace(')',''))/38,0)+1)
        except:
            count = 5
        print(f'Puslapiai {count}')
        for i in range(count):
            try:
                page = i + 1
                url = f'{urlas}?p={page}'
                driver.get(url)
                source = driver.page_source
                bs = BeautifulSoup(source, 'html.parser')
                dviraciai = bs.find_all('div', {'class':'GRID_ITEM'})
                for dv in dviraciai:
                    try:
                        zod = {}
                        zod['Skelbimo_tipas'] = urlas
                        x = dv.find('a')
                        # print(x['href'])
                        url = f'https://www.varle.lt{x['href']}'
                        # if True:
                        if not check_if_value_exists_or_max('../../../web_scrap.db','varlelt','URL',url):
                            zod['URL'] = url
                            driver.get(url)
                            source = driver.page_source
                            ss = BeautifulSoup(source, 'html.parser')
                            price = ss.find('span', {'class':'price-value'}).text.strip()
                            zod['Kaina'] = price
                            fspec = ss.find_all('div', {'class':'PRODUCT_SPEC'})
                            for s in fspec:
                                try:
                                    specif = s.find_all('li', {'class':'spec-line'})
                                    for sp in specif:
                                        try:
                                            x = None
                                            y = None
                                            x = sp.find('div', {'class':'left-column'}).text.strip()
                                            y = sp.find('div', {'class':'right-column'}).text.strip()
                                            zod[x]=y
                                        except:
                                            None
                                except:
                                    None
                            spec = ss.find_all('div', {'class':'PRODUCT_SPEC_FILTERS'})
                            for s in spec:
                                try:
                                    specif = s.find_all('li', {'class':'spec-line'})
                                    for sp in specif:
                                        try:
                                            x = None
                                            y = None
                                            x = sp.find('span', {'class':'left-column'}).text.strip()
                                            y = sp.find('span', {'class':'right-column'}).text.strip()
                                            zod[x]=y
                                        except:
                                            None
                                except:
                                    None
                            insert_into_db(zod,'../../../web_scrap.db','varlelt')
                    except:
                        None
            except:
                None
    except Exception as e:
        print(e)
driver.close()
```
