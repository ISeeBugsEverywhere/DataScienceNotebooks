```python
regs =  pd.read_csv('C:/Users/manta/OneDrive/Dokumentai/Python projektai/VietovesTemperaturaDregme.csv')
```

```python
regs.head()
```

```python
regs['laikas'] = pd.to_datetime(regs['laikas'])
```

```python
fig = px.scatter_mapbox(data_frame=regs, lon='ilguma', lat='platuma', zoom=6.1)
fig.update_layout(mapbox_style='open-street-map')
fig.update_layout(width=800, height=600)
fig.show()
```

```python
regs_id = list(set(list(regs['vietoves_registratoriaus_id'])))
```

```python
reg1 = regs_id[1]
reg1df = regs[regs['vietoves_registratoriaus_id']==reg1]
```

```python
print(reg1)
```

```python
fig, axes = plt.subplots(figsize=(10,6))

axes.set_title(f'Temperatūros registratoriaus {reg1}')
sns.lineplot(data=reg1df,x = 'laikas', y='temperatura_8cm_gylyje_dirvozemyje', ax=axes,label='-8cm', zorder=5)
sns.lineplot(data=reg1df,x = 'laikas', y='temperatura_15cm_virs_dirvozemio', ax=axes,label='15cm')
sns.lineplot(data=reg1df,x = 'laikas', y='temperatura_dirvozemio_pavirsiuje', ax=axes,label='0cm')
axes.set(xlabel='Data ir laikas',ylabel='Temperatūra')
axes.legend()
plt.show()
```
