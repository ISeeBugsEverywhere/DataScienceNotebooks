```python
# bar'ai. ir horizontalūs bar'ai

a = np.arange(1,4)
b = np.random.randint(2,15,3)
c = np.random.randint(2,15,3)
labels = ['Vilnius', 'Kaunas', 'Klaipėda']

fig, axis = plt.subplots()
w = 0.25
P = axis.bar(a-w/2,b, label='A', color='xkcd:deep purple', width=w)
B = axis.bar(a+w/2,c, label='B', color='xkcd:cyan', width=w)

axis.bar_label(P)
axis.bar_label(B)

axis.set_xticks(a)
axis.set_xticklabels(labels)
axis.legend()
plt.show()
```

```python
#nuskaityti ir atvaizduoti fpc.csv failo stulpelius grafiškai 
from dateutil import parser
class TxtReader():
    '''
    Nuskaito failą, tekstinį
    konvertuoja į skaičius,
    pirmą stulpelį gali
    konvertuoti į datetime
    '''
    def __init__(self, fname, sep=',', use_xy=[0,1],skip=0, convert_date=False) -> None:
        self.fname= fname
        self.skip = skip
        self.xy = use_xy
        self.sep=sep
        self.convert = convert_date
        self.__X = None
        self.__Y = None
        self.read_file()
        
    
    def read_file(self):
        f = open(self.fname, mode='r', encoding='utf-8')
        lines = f.readlines()
        f.close()
        xl = []
        yl = []
        for line in lines[self.skip:]:
            r = line.split(self.sep)
            xi = r[self.xy[0]]
            yi = r[self.xy[1]]
            if self.convert:
                x = parser.parse(xi)
                xl.append(x)
            else:
                x = float(xi)
                xl.append(x)
            y = float(yi)
            yl.append(y)
        self.__X = xl
        self.__Y = yl
    
    def X(self):
        return self.__X
    def Y(self):
        return self.__Y
            
```
