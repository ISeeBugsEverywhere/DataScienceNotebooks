```python

# • Parašykite klasę, turinčią konstruktorių. Konstruktorius turi
# spausdinti pranešimą, jog objektas yra sukurtas.
class SimpleObject():
    def __init__(self) -> None:
        self.X = "Tai yra atributas"        
r = SimpleObject()
atributas_x = r.X
print(atributas_x)
# • Parašykite klasę, kuri turėtų konstruktorių, destruktorių. Abu
# turi spausdinti informacinį pranešimą, jog objektas yra
# sukurtas/sunaikintas.
class SimpleObject():
    def __init__(self):
        print( "Tai yra atributas"   )
    def __del__(self) -> None:
        print( "Atributas sunaikintas"    )
b = SimpleObject()
del b


# • Parašykite klasę, kuri turėtų vieną atributą. Pasiekite jo
# reikšmę, išveskite. Pakeiskite atributo reikšmę. Patikrinkite,
# ar reikšmė pakito.
class SimpleObject():
    def __init__(self, x) -> None:
        self.X = x
        self.XX = x
        
    def get_X(self): #metodas
        return self.X #atributo x pasiekimas metodo viduje
    
    def set_X(self,x):
        self.X = x
        
r = SimpleObject(42)
print(r.X)

r.set_X(88)

print(r.get_X())
if r.get_X() != r.XX:
    print("Reiksmė pasikeitė",r.X)
else:
    print("Reikšmė nepasikeitė",r.XX)

# • Prie ankstesnės klasės pridėkite metodus, skirtus išvesti
# atributo reikšmę, pakeisti atributo reikšmę. patikrinkite, ar jie
# veikia.
```

```python
import time
class Tarakonas():
    def __init__(self) -> None:
        print('išsirito tarakonas')
        self.gimtadienis=time.time()
    def __del__(self):
        print('Tarakonas mirė. Jis gyveno:',time.time()-self.gimtadienis)
Rudis=Tarakonas()
time.sleep(0.7)
Juodis=Tarakonas()
time.sleep(0.7)
Pilkis=Tarakonas()

print(Rudis.gimtadienis)
print(Juodis.gimtadienis)
print(Pilkis.gimtadienis)

del Rudis
del Juodis
del Pilkis
```

```python
# Perrašyti ankstesnę funkciją taip, kad jai galima būtų nurodyti
# stulpelių skirtuką, ir kuriuos stulpelius grąžinti kaip sąrašus. Stulpeliai nurodomi kaip
# indeksai (skaičiai).
# Turi veikti ir su duotaisiais DAT ir su duotaisiais txt failais.
# Taip pat funkcijai turite galėti nurodyti, koks
# skirtukas skiria dešimtaines dalis - taškas ar kablelis.
# funkcija turi apdoroti tik tas eilutes, kur nurodytame pirmąjame stulpelyje 
# yra skaičiai, turi ignoruoti tekstinius įrašus

f_name = str(input('Iveskite failo pavadinima'))
skir = str(input('nurodykite skirtuka'))
if skir == '':
    skir = ' '

def scan(x,c):
    
    f = open(x, mode='r', encoding='utf-8')


    turinys = f.readlines()[2:-4]
    s1 = []
    s2 = []
    s3 = []
    s4 = []
    s5 = []
    s6 = []

    a = []
    
    for i in turinys:


        a.append(i.split(c))
    # print(a)
    for s in a:
        if s.split()[0] == 'Tamsa':
            pass
        else:
                
            s1.append(float(s[0]))
            s2.append(float(s[1]))
            s3.append(float(s[2]))
            s4.append(float(s[3]))
            s5.append(float(s[4]))
            s6.append(s[5])
            
    print(s1,s2,s3,s4,s5,s6,sep='\n')

    f.close()
    
scan(f_name,skir)
```
