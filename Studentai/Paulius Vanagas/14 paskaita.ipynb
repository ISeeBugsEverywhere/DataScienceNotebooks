```python
# A) Duotas sąrašas x=[1,2,3,4,5,6,7,8,9]
# Viename lange, skirtingose ašyse atvaizduokite x2 , x ⋅ 3,
# x ⋅ a, čia a - įvedamas vartotojo.
# Panaudokite skirtingas spalvas, linijų tipus.
# B) Duoti sąrašai x = [1,2,3,4,5], y1=[2,2,0,0,2], y2 =
# [4,3,2,1,-1], y3 = [2,4,9,16,25], y4 = [-1,1,-1,1,-1]
# Atvaizduokite viename lange x~y1, x~y2,
# x~y3, x~y4 kreives (grafikus). Pirmasis grafikas - linijinis, antrasis - taškinis,
# trečiasis - linija su duomenų taškais, ketvirtasis - brūkšninis.
# Spalvos visų turi būti skirtingos. Grafikai, ašys turi turėti
# pavadinimus.

# Sugeneruoti sąrašą x, turintį 101 elementą (nuo 0 iki 100).
# Sukurti antrą sąrašą, kuriame būtų skaičiai, pakelti kvadratu, iš
# pirmojo sąrašo (x2 )
# Sukurti trečiąjį sąrašą, kuriame skaičiai būtų pakelti kvadratu ir
# padauginti iš atsitiktinai sugeneruoto skaičiaus (x2 ⋅ a).
# Sugeneruoti 101-to elementų ilgio sąrašą iš atsitiktinių skaičių.
# Visus šiuos sąrašus atvaizduoti grafike. Grafikas turi turėti
# pavadinimą, pavadintos ašys, pakeisti šriftų dydžiai.
```

```python
x=[1,2,3,4,5,6,7,8,9]
a= -0.5

fig, axis = plt.subplots(1, 3, figsize=(12, 4))

axis[0].plot(x, list(map(lambda x: x**2, x)), marker='s', label='pirmas')
axis[1].plot(x, list(map(lambda x: x*3, x)), linewidth=3, color='#fe01b1', linestyle=':', label='trecias')
axis[2].plot(x, list(map(lambda x: x*a, x)), linewidth=8, color='xkcd:leaf green', linestyle='--', label='tingiu')

axis[0].legend(loc='best')
axis[1].legend(loc='best')
axis[2].legend(loc='best')

fig.tight_layout()
plt.show()
```

```python
x = [1,2,3,4,5]
y1=[2,2,0,0,2]
y2 =[4,3,2,1,-1]
y3 = [2,4,9,16,25]
y4 = [-1,1,-1,1,-1]

fig, axis = plt.subplots()

axis.plot(x, y1, label='pirmas')
axis.plot(x, y2,  linestyle='dotted', label='antras')
axis.plot(x, y3, marker='o', label='trecias')
axis.plot(x, y4,  linestyle='--', label='ketvirtas')
axis.legend(loc='best')

axis.set_title('bandymukas')
axis.set_xlabel('x asiuke')
axis.set_ylabel('ygriukas')

fig.tight_layout()
plt.show()
```

```python
x = np.arange(0, 101)

x2= list(map(lambda z: z**2, x))

x2a= list(map(lambda z: z*(np.random.rand()), x2))

x101=np.random.randint(0, 5000, 101)


fig, axis = plt.subplots()

axis.plot(x, x2, label='kvadratu')
axis.plot(x, x2a, label='x2a')
axis.plot(x, x101, label='atsitiktiniai')
axis.legend(loc='best')

axis.set_title('bandymukas')
axis.set_xlabel('x asiuke')
axis.set_ylabel('ygriukas')

fig.tight_layout()
plt.show()
```
