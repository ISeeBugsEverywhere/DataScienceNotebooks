```python
def PieOfPie(pie1=[], pie2=[], pie1_labels=[], pie2_labels=[], isskleist=0):
    if isskleist != 0:
        pie1.append(pie1[0])
        pie1[0]=pie1[isskleist]
        pie1.pop(isskleist)
        pie1_labels.append(pie1_labels[0])
        pie1_labels[0]=pie1_labels[isskleist]
        pie1_labels.pop(isskleist)
        
    # make figure and assign axis objects
    fig = plt.figure(figsize=(9, 5))
    ax1 = fig.add_subplot(1, 2, 1)
    ax2 = fig.add_subplot(1, 2, 2)
    fig.subplots_adjust(wspace=0.2)

    # large pie chart parameters
    explode = [0.1]
    i=1
    while i < len(pie1):
        explode.append(0)
        i=i+1
        
    # rotate so that first wedge is split by the x-axis
    angle = -180*(pie1[0]/sum(pie1))
    ax1.pie(pie1, autopct='%1.1f%%', startangle=angle,    
        labels=pie1_labels, explode=explode)

    # small pie chart parameters
    width = .2
    ax2.pie(pie2, autopct='%1.1f%%', startangle=angle,    
        labels=pie2_labels, radius=0.5, textprops={'size': 'smaller'})

    ax1.set_title('Mėgstamiausia spalva')
    ax2.set_title('Amžiaus grupė')

    # use ConnectionPatch to draw lines between the two plots
    # get the wedge data
    theta1, theta2 = ax1.patches[0].theta1, ax1.patches[0].theta2
    center, r = ax1.patches[0].center, ax1.patches[0].r

    # draw top connecting line
    x = r * np.cos(np.pi / 180 * theta2) + center[0]
    y = np.sin(np.pi / 180 * theta2) + center[1]
    con = ConnectionPatch(xyA=(- width / 2, .5), xyB=(x, y),
                      coordsA="data", coordsB="data", axesA=ax2, axesB=ax1)
    con.set_color([0, 0, 0])
    con.set_linewidth(2)
    ax2.add_artist(con)

    # draw bottom connecting line
    x = r * np.cos(np.pi / 180 * theta1) + center[0]
    y = np.sin(np.pi / 180 * theta1) + center[1]
    con = ConnectionPatch(xyA=(- width / 2, -.5), xyB=(x, y), coordsA="data",
                        coordsB="data", axesA=ax2, axesB=ax1)
    con.set_color([0, 0, 0])
    ax2.add_artist(con)
    con.set_linewidth(2)

    plt.show()


spalvos_daznis=[150, 140, 175, 188, 120]
spalvos =['zalia', 'melyna', 'geltona', 'juoda', 'pilka']
isskleidziama_spalva= 1
amzius =['iki 20', 'nuo 21 iki 50', 'nuo 51']
amzius_kiekis=[70, 40, 40]
PieOfPie(spalvos_daznis, amzius_kiekis, spalvos, amzius, isskleidziama_spalva)
```
