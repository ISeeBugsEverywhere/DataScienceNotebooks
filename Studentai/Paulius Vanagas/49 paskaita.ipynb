```python
# Užduotis 1: Funkcija, kuri tikrina, ar skaičius yra nelyginis
def arNelyginis(x:int) -> bool:
    '''sums up two digits
    >>> arNelyginis(-21)
    True
    >>> arNelyginis(5)
    True
    >>> arNelyginis(6)
    False
    >>> arNelyginis(3.5)
    False
    '''
    if isinstance(x, int) and x % 2 != 0  :
        return True
    else:
        return False

# Užduotis 2: Funkcija, kuri tikrina, ar skaičius yra teigiamas
def arTeigiamas(x: float|int) -> bool:
    '''sums up two digits
    >>> arTeigiamas(-21)
    False
    >>> arTeigiamas('as')
    False
    >>> arTeigiamas(6.5)
    True
    >>> arTeigiamas(0)
    False
    '''
    if isinstance(x, float|int) and x >0:
        return True
    else:
        return False
    

# Užduotis 3: Funkcija, kuri tikrina, ar skaičius yra neigiamas
def arNeigiamas(x: float|int) -> bool:
    '''sums up two digits
    >>> arNeigiamas(-21)
    True
    >>> arNeigiamas('as')
    False
    >>> arNeigiamas(6.5)
    False
    >>> arNeigiamas(0)
    False
    '''
    if isinstance(x, float|int) and x <0:
        return True
    else:
        return False
    
# Užduotis 4: Funkcija, kuri tikrina, ar tekstas yra tuščias
def arTuscias(x:str) -> bool:
    '''sums up two digits
    >>> arTuscias('')
    True
    >>> arTuscias('as')
    False
    >>> arTuscias(6.5)
    False
    >>> arTuscias(' ')
    False
    '''
    if isinstance(x, str) and len(x)==0:
        return True
    else:
        return False


# Užduotis 5: Funkcija, kuri tikrina, ar tekstas prasideda tam tikra raide
def arRaide(x:str, y:str) -> bool:
    '''sums up two digits
    >>> arRaide('r', 'r asd dasd')
    True
    >>> arRaide('R', 'r asd dasd')
    False
    >>> arRaide(6.5, '6asd 6asd')
    False
    >>> arRaide('', ' asd a')
    False
    '''
    if isinstance(x, str) and isinstance(y, str) and x==y[0]:
        return True
    else:
        return False
    
# Užduotis 6: Funkcija, kuri suranda pateikto skaičių sąrašo vidurkį (len, sum naudoti viduje)
def skaiciuVidurkis(x:list) -> float:
    '''sums up two digits
    >>> skaiciuVidurkis(15)
    'turi būti pateiktas skaičių sąrašas'
    >>> skaiciuVidurkis([12, 12, 12])
    12.0
    >>> skaiciuVidurkis([2, 5])
    3.5
    >>> skaiciuVidurkis([12, 'fgh'])
    'turi būti pateiktas skaičių sąrašas'
    '''
    try: 
        vidurkis= sum(x)/len(x)
    except: 
        return 'turi būti pateiktas skaičių sąrašas'
    else:
        return vidurkis
    

# Funkcija grąžina didžiausią sąrašo elementą.
def didziausiasElementas(x:list) -> float:
    '''sums up two digits
    >>> didziausiasElementas(15)
    'blogi duomenys'
    >>> didziausiasElementas([0, -5, 12])
    12
    >>> didziausiasElementas([-6, -6, -6, -6])
    -6
    >>> didziausiasElementas([12, 'a', -5])
    Traceback (most recent call last):
    TypeError: '>' not supported between instances of 'str' and 'int'
    '''
    if isinstance(x, list) and isinstance(max(x), int):
        return max(x)
    else:
        return 'blogi duomenys'


# Funkcija apverčia duotą eilutę.
def apvercia(x:str) -> str:
    '''sums up two digits
    >>> apvercia('15')
    '51'
    >>> apvercia('')
    ''
    >>> apvercia('0 1 2')
    '2 1 0'
    '''
    return x[::-1]


# Funkcija apskaičiuoja faktorialą.
def faktorialas(x:int) -> int:
    '''sums up two digits
    >>> faktorialas(0)
    1
    >>> faktorialas(1)
    1
    >>> faktorialas(3)
    6
    >>> faktorialas(5.3)
    'įvestas skaičius turi būti natūralusis'
    '''
    f=1
    if isinstance(x, int) and x >= 0:
        for i in range(x):
            f=(i+1)*f
        return f
    else: return 'įvestas skaičius turi būti natūralusis'
    

# Funkcija grąžina Fibonacci seką iki n-tosios reikšmės.
def Fibonacci(x:int) -> list:
    '''sums up two digits
    >>> Fibonacci(1)
    [0]
    >>> Fibonacci(2)
    [0, 1]
    >>> Fibonacci(0)
    'blogi duomenys'
    >>> Fibonacci(5.3)
    'blogi duomenys'
    '''
    f=[0, 1]
    if x == 1:
        return f[0:1]
    elif x==2:
        return f
    elif isinstance(x, int) and x > 2:
        i=3
        while i <= x:
            f.append(f[-1]+f[-2])
            i+=1
        return f
    else:
        return 'blogi duomenys'

        
if __name__ == '__main__':
    doctest.testmod(verbose=True, globs=globals(), report=True)
```
