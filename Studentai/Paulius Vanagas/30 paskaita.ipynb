```python
def keitimas(a):
    if 0<= a <= 6 or a == 23:
        return 'naktis'
    if 7<= a <= 10:
        return 'rytas'
    if 11<= a <= 12:
        return 'priešpietė'
    if 13<= a <= 14:
        return 'pietūs'
    if 15<= a <= 18:
        return 'pavakarė'
    if 19<= a <= 22:
        return 'vakaras'
    

times = np.random.randint(0,24,1000)
laikai = list(map(keitimas, times))

tadam = {'naktis':0, 'rytas':0, 'priešpietė': 0, 'pietūs': 0, 'pavakarė': 0, 'vakaras': 0}
for i in laikai:
    if i == 'naktis':
        tadam['naktis']=tadam['naktis']+1
    if i == 'rytas':
        tadam['rytas']=tadam['rytas']+1
    if i == 'priešpietė':
        tadam['priešpietė']=tadam['priešpietė']+1
    if i == 'pietūs':
        tadam['pietūs']=tadam['pietūs']+1
    if i == 'pavakarė':
        tadam['pavakarė']=tadam['pavakarė']+1
    if i == 'vakaras':
        tadam['vakaras']=tadam['vakaras']+1
        
print(tadam)

fig, axis = plt.subplots()

a=axis.bar(tadam.keys(), tadam.values())
axis.bar_label(a)

plt.show()

# destytojo patarimas del hard codinimo vengimo: set(), list(), dict.fromkeys()
# .count()
```

```python
def keitimas(a):
    if 0<= a <= 6 or a == 23:
        return 'naktis' 
    if 7<= a <= 10:
        return 'rytas'
    if 11<= a <= 12:
        return 'priešpietė'
    if 13<= a <= 14:
        return 'pietūs'
    if 15<= a <= 18:
        return 'pavakarė'
    if 19<= a <= 22:
        return 'vakaras'
    

times = np.random.randint(0,24,1000)
laikai = list(map(keitimas, times))

laikusarasas=list(set(laikai))
tadam={}
for i in laikusarasas:
    tadam[i]=0

for i in laikai:
        tadam[i]=tadam[i]+1
        
print(tadam)

fig, axis = plt.subplots()

a=axis.bar(tadam.keys(), tadam.values())
axis.bar_label(a)

plt.show()

# destytojo patarimas del hard codinimo vengimo: set(), list(), dict.fromkeys()
# .count()
```
