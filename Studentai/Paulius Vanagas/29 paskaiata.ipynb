```python
nypd['V'] = nypd['DATA'].dt.hour
nypd.head(3)
```

```python
nypdII = nypd[['DATA', 'V', 'BORO', 'STATISTICAL_MURDER_FLAG']]
nypdII.head(3)
```

```python
nypd_boro = nypdII.groupby('BORO')

boros = nypd_boro.groups.keys()

fig, axis = plt.subplots()

for boro in boros:
    b = nypd_boro.get_group(boro)
    h = b['V'].value_counts().sort_index(ascending= True)
    
    axis.plot(h.index, h.values, label = boro, marker ='o')

axis.set_xticks(h.index)
axis.legend(loc='best')
plt.show()
    
```

```python
# Atvaizduokite grafiką, kuriame būtų tik mirtinų susiškaudymų kiekiai per valandą kiekviename rajone. Panaudokite plot() grafiką.
# Su pie plot atvaizduokite, kiek mirtinų nusišaudymų tenka kiekvienam rajonui.
```

```python

nypd_mirtini = nypdII[nypdII['STATISTICAL_MURDER_FLAG']== True]
mirtini = nypd_mirtini.groupby('BORO')
boros = mirtini.groups.keys()

fig, axis = plt.subplots()

visos_valandos = np.arange(0, 24)

for boro in boros:
    b = mirtini.get_group(boro)
    h = b['V'].value_counts().sort_index(ascending= True)
    
    h = h.reindex(visos_valandos, fill_value=0)
    
    axis.plot(h.index, h.values, label = boro, marker ='o')

axis.set_xticks(h.index)
axis.legend(loc='best')
plt.show()
```

```python
dus2018 = pd.read_csv('../../DATA/DUS2018.csv')[['amzius', 'lytis', 'bdu_spalio']]
dus2018.head(2)
```

```python
vid=dus2018['bdu_spalio'].mean()

dd2018F = dus2018.query('bdu_spalio >=  @vid and lytis== "F"')
dd2018M = dus2018.query('bdu_spalio >=  @vid and lytis== "M"')

mot=dd2018F['amzius'].value_counts().sort_index()
vyr=dd2018M['amzius'].value_counts().sort_index()

plt.pie(x = mot.values, labels = mot.index, autopct='%.2f%%')
plt.title('moterys')
plt.show()

plt.pie(x = vyr.values, labels = vyr.index, autopct='%.2f%%')
plt.title('vyrai')
plt.show()
```
