```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestRegressor
# Load dataset
data = load_iris()
X, y = data.data, data.target
# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
# Initialize and train the RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
# Make predictions
y_pred = rf.predict(X_test)
# Evaluate accuracy
accuracy = accuracy_score(y_test, y_pred) 
print(f'Accuracy:{accuracy}')
```

```python
sar = df['air_temperature'].to_list()
Tskirtumas=[0]

for i in range(0, len(df)-1):
    Tskirtumas.append(round(sar[i]-sar[i+1], 2))
```

```python
df['temp_skirtumas']= Tskirtumas
```

```python
df['obs_time_utc'] = df['obs_time_utc'].apply(lambda x: int(x.split(' ')[1].split(':')[0]))
```

```python
df['condition_code'] = LabelEncoder().fit_transform(df['condition_code'])
condition_code = dict(zip(df['condition_code'], df['condition_code']))
```

```python
df=df.dropna()
```

```python
X= df.drop(columns=['air_temperature', 'feels_like_temperature', 'temp_skirtumas', 'condition_code' ])
y=df['air_temperature']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=37)

fit = LinearRegression().fit(X_train, y_train)

pred = fit.predict(X_test)
fit.score(X_test, y_test)*100.0
```

```python
pred = [round(num, 2) for num in pred]
py_list = [float(num) for num in pred]
print(py_list)
```

```python
rf = RandomForestRegressor(n_estimators=70, random_state=35)
rf.fit(X_train, y_train)
# Make predictions
y_pred = rf.predict(X_test)
# Evaluate accuracy
mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
print(f'Mean Absolute Percentage Error: {mape}%')
```
