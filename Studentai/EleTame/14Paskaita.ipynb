```python
#A) Duotas sraas x=[1,2,3,4,5,6,7,8,9]
#Viename lange, skirtingose ayse atvaizduokite x2 , x  3,
#x  a, ia a - vedamas vartotojo.
#Panaudokite skirtingas spalvas, linij tipus

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6, 7, 8, 9]

# Apskaiiuojame y reikmes
y_kv = [i**2 for i in x]
y_triple = [i * 3 for i in x]

# Praome vartotojo vesti 'a' reikm
a = float(input("veskite skaii a: "))
y_a = [i * a for i in x]

# Sukuriame grafikus
fig, axis = plt.subplots(2, 2, figsize=(12, 8))

# x^2 grafikas
axis[0, 0].plot(x, y_kv, color='green', linestyle='-.', marker='s', label='x^2', linewidth=2, markersize=24)
axis[0, 0].set_title('x^2')
axis[0, 0].legend()

# x * 3 grafikas
axis[0, 1].plot(x, y_triple, color='red', linestyle='--', marker='o', label='x * 3',linewidth=4, markersize=16)
axis[0, 1].set_title('x * 3')
axis[0, 1].legend()

# x * a grafikas
axis[1, 0].scatter(x, y_a, color='blue', marker='h', label=f'x * {a}', s=1000)
axis[1, 0].set_title(f'x * {a}')
axis[1, 0].legend()

# Paaliname paskutin a
fig.delaxes(axis[1, 1])

# Pridedame bendrus pavadinimus ir tinklelius
for ax in axis.flat:
    ax.set(xlabel='x values', ylabel='y values')
    ax.label_outer()
    ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
#A) Duotas sraas x=[1,2,3,4,5,6,7,8,9]
#Viename lange, skirtingose ayse atvaizduokite x2 , x  3,
#x  a, ia a - vedamas vartotojo.
#Panaudokite skirtingas spalvas, linij tipus
#CIA KOMPAKTISKESNIS KODAS. KODE MAZIAU EILUCIU. NAudojama lambda f-ja.

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6, 7, 8, 9]

# Praome vartotojo vesti 'a' reikm
a = float(input("veskite skaii a: "))


# Sukuriame grafikus
fig, axis = plt.subplots(2, 2, figsize=(12, 8))

# x^2 grafikas
axis[0, 0].plot(x, list(map(lambda x: x**2, x)),  color='green', linestyle='-.', marker='s', label='x^2', linewidth=6, markersize=6)
axis[0, 0].set_title('x^2')
axis[0, 0].legend()

# x * 3 grafikas
axis[0, 1].plot(x, list(map(lambda x: x*3, x)), color='red', linestyle='--', marker='o', label='x * 3',linewidth=4, markersize=12)
axis[0, 1].set_title('x * 3')
axis[0, 1].legend()

# x * a grafikas
axis[1, 0].scatter(x, list(map(lambda x: x*a, x)), color='blue', marker='h', label=f'x * {a}', s=100)
axis[1, 0].set_title(f'x * {a}')
axis[1, 0].legend()

# Paaliname paskutin a
fig.delaxes(axis[1, 1])

# Pridedame bendrus pavadinimus ir tinklelius
for ax in axis.flat:
    ax.set_xlabel('x values', fontsize=16)# Nurodant fontu dydi, reikia atskirai asis aprasyti
    ax.set_ylabel('y values', fontsize=16)
    ax.label_outer()
    ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Duoti sraai  = [1,2,3,4,5], y1=[2,2,0,0,2], y2 =
#[4,3,2,1,-1], y3 = [2,4,9,16,25], y4 = [-1,1,-1,1,-1]
#Atvaizduokite viename lange x~y1, x~y2,
#x~y3, x~y4 kreives (grafikus). Pirmasis grafikas - linijinis, antrasis - takinis,
#treiasis - linija su duomen takais, ketvirtasis - brkninis.
#Spalvos vis turi bti skirtingos. Grafikai, ays turi turti
#pavadinimus.
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 =[2, 2, 0, 0, 2]
y2 =[4, 3, 2, 1,-1]
y3 =[2, 4, 9, 16, 25]
y4 =[-1, 1,-1, 1,-1]


# Sukuriame grafikus
# Sukuriame grafikus
fig, axis = plt.subplots(2, 2, figsize=(12, 8))

# y1 grafikas
axis[0, 0].plot(x, y1, color='green', linestyle='-', marker='s', label='y1', linewidth=2, markersize=8)
axis[0, 0].set_title('y1')
axis[0, 0].legend(loc='best')

# y2 grafikas
axis[0, 1].scatter(x, y2, color='red', marker='o', label='y2', s=100)
axis[0, 1].set_title('y2')
axis[0, 1].legend()

# y3 grafikas
axis[1, 0].plot(x, y3, color='black', linestyle='-', marker='x', label='y3', linewidth=2, markersize=8)
axis[1, 0].set_title('y3')
axis[1, 0].legend()

# y4 grafikas
axis[1, 1].plot(x, y4, color='blue', linestyle='-.', marker='h', label='y4', linewidth=2, markersize=8)
axis[1, 1].set_title('y4')
axis[1, 1].legend()

# Pridedame bendrus pavadinimus ir tinklelius
for ax in axis.flat:
    ax.set(xlabel='x values', ylabel='y values')
    ax.label_outer()
    ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
#Sugeneruoti sra x, turint 101 element (nuo 0 iki 100).
#Sukurti antr sra, kuriame bt skaiiai, pakelti kvadratu, i
#pirmojo srao (x2 )
#Sukurti treij sra, kuriame skaiiai bt pakelti kvadratu ir
#padauginti i atsitiktinai sugeneruoto skaiiaus (x2  a).
#Sugeneruoti 101-to element ilgio sra i atsitiktini skaii.
#Visus iuos sraus atvaizduoti grafike. Grafikas turi turti
#pavadinim, pavadintos ays, pakeisti rift dydiai.
import matplotlib.pyplot as plt
import random

# Sugeneruoti sra x, turint 101 element (nuo 0 iki 100)
x = list(range(0, 101))

# Sukurti antr sra, kuriame bt skaiiai, pakelti kvadratu
x2 = [i**2 for i in x]

# Sugeneruoti atsitiktin skaii
a = random.random()

# Sukurti treij sra, kuriame skaiiai bt pakelti kvadratu ir padauginti i atsitiktinio skaiiaus
x2_a = [i**2 * a for i in x]

# Sugeneruoti 101-to element ilgio sra i atsitiktini skaii
random_list = [random.random() for _ in x]

# Sukuriame grafikus
plt.figure(figsize=(14, 8))

# x vs x2
plt.plot(x, x2, label='x^2', color='blue')

# x vs x2_a
plt.plot(x, x2_a, label=f'x^2 * {a:.2f}', color='green')

# x vs random_list
plt.plot(x, random_list, label='Random values', color='red')

# Pridedame pavadinimus ir etiketes
plt.title('Graphs of x^2, x^2 * a, and Random Values', fontsize=20)
plt.xlabel('x values', fontsize=15)
plt.ylabel('y values', fontsize=15)
plt.legend(fontsize=12)

# Rodyti grafikus
plt.grid(True)
plt.show()
```
