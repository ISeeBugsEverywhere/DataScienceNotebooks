```python
#times = np.random.randint(0,24,1000) intervalai : 0-6 - naktis, 7-10 - rytas,
# 11-12 - priešpietė, 13-14-pietūs, 15-18-pavakarė, 19-22 - vakaras, 
# 23 - 0 - naktis. 
#Parašykite funkciją, kuri iš times sąrašo verčių sugeneruotų tekstų 
# (rytas, vakaras, popietė ... ) sąrašą. Pritaikykite map(). 
# Suskaičiuokite, kiek kartu pasikartojo 
#kiek kartų pasikartojo atitinkami tekstai, atvaizduokite tai su bar arba barh.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sugeneruojame atsitiktinius laikus
times = np.random.randint(0, 24, 1000)

# Funkcija, kuri pagal laiką priskiria intervalą
def get_interval(hour):
    if 0 <= hour <= 6 or hour == 23:
        return 'naktis'
    elif 7 <= hour <= 10:
        return 'rytas'
    elif 11 <= hour <= 12:
        return 'priešpietė'
    elif 13 <= hour <= 14:
        return 'pietūs'
    elif 15 <= hour <= 18:
        return 'pavakarė'
    elif 19 <= hour <= 22:
        return 'vakaras'
    else:
        return 'nežinoma'

# Priskiriame laikų intervalus
times_intervals = list(map(get_interval, times))

# Sukuriame DataFrame su intervalais
df = pd.DataFrame({'intervalas': times_intervals})

# Suskaičiuojame, kiek kartų pasikartojo kiekvienas intervalas
interval_counts = df['intervalas'].value_counts()

# Duomenų paruošimas grafikui
labels = interval_counts.index
counts = interval_counts.values

# Braižome grafiką
plt.figure(figsize=(10, 6))
plt.bar(labels, counts, color='lightgreen')
plt.xlabel('Laiko intervalai')
plt.ylabel('Pasikartojimų skaičius')
plt.title('Laiko intervalų pasikartojimai')

# Pridėti etiketes su pasikartojimų skaičiumi prie stulpelių
for i in range(len(counts)):
    plt.text(i, counts[i] + 5, str(counts[i]), ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
#prisiminti .dt. modulį pandos:
# prie nypd pridėkite naują stulpelį V, kur būtų tik įvykio valanda
# tada pritaikykite savo rašytą funkciją su .apply() ir sukurkite naują
# stulpelį su tekstais rytas, vakaras, priešpietė
# tada vizualizuokite, kuriuo paros metu kiek buvo susišaudymų (bar arba barh)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Nuskaitome failą ir sujungiame datą bei laiką
nypd = pd.read_csv('../../DATA/NYPD.csv', parse_dates=[['OCCUR_DATE', 'OCCUR_TIME']])

# Pridedame naują stulpelį V, kuris ištraukia įvykio valandą
nypd['V'] = nypd['OCCUR_DATE_OCCUR_TIME'].dt.hour

# Funkcija, kuri pagal valandą priskiria intervalą
def get_interval(hour):
    if 0 <= hour <= 6 or hour == 23:
        return 'naktis'
    elif 7 <= hour <= 10:
        return 'rytas'
    elif 11 <= hour <= 12:
        return 'priešpietė'
    elif 13 <= hour <= 14:
        return 'pietūs'
    elif 15 <= hour <= 18:
        return 'pavakarė'
    elif 19 <= hour <= 22:
        return 'vakaras'
    else:
        return 'nežinoma'

# Pritaikome funkciją su .apply() naujai sukurtam stulpeliui
nypd['laiko_intervalas'] = nypd['V'].apply(get_interval)

# Suskaičiuojame, kiek kartų pasikartojo kiekvienas laiko intervalas
interval_counts = nypd['laiko_intervalas'].value_counts()

# Duomenų paruošimas grafikui
labels = interval_counts.index
counts = interval_counts.values

# Braižome bar diagramą
plt.figure(figsize=(10, 6))
plt.bar(labels, counts, color='lightgreen')
plt.xlabel('Laiko intervalai')
plt.ylabel('Pasikartojimų skaičius')
plt.title('Susišaudymų pasikartojimai pagal paros laiką')

# Pridedame etiketes prie stulpelių
for i in range(len(counts)):
    plt.text(i, counts[i] + 5, str(counts[i]), ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
#prisiminti .dt. modulį pandos:
# prie nypd pridėkite naują stulpelį V, kur būtų tik įvykio valanda
# tada pritaikykite savo rašytą funkciją su .apply() ir sukurkite naują
# stulpelį su tekstais rytas, vakaras, priešpietė
# tada vizualizuokite, kuriuo paros metu kiek buvo susišaudymų (bar arba barh)
#Isdestau paros laika tvarkingai


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.api.types import CategoricalDtype

# Nuskaitome failą, paliekame tik  datą bei laiką, kuriuos sujungiame
nypd = pd.read_csv('../../DATA/NYPD.csv', parse_dates=[['OCCUR_DATE', 'OCCUR_TIME']])

# Pridedame naują stulpelį V, kuris ištraukia įvykio valandą
nypd['V'] = nypd['OCCUR_DATE_OCCUR_TIME'].dt.hour

# Funkcija, kuri pagal valandą priskiria intervalą
def get_interval(hour):
    if 0 <= hour <= 6 or hour == 23:
        return 'naktis'
    elif 7 <= hour <= 10:
        return 'rytas'
    elif 11 <= hour <= 12:
        return 'priešpietė'
    elif 13 <= hour <= 14:
        return 'pietūs'
    elif 15 <= hour <= 18:
        return 'pavakarė'
    elif 19 <= hour <= 22:
        return 'vakaras'
    else:
        return 'nežinoma'

# Pritaikome funkciją su .apply() naujai sukurtam stulpeliui
nypd['laiko_intervalas'] = nypd['V'].apply(get_interval)

# Nustatome pasirinktos eiliškumo tvarkos kategoriją
category_order = ['naktis', 'rytas', 'priešpietė', 'pietūs', 'pavakarė', 'vakaras']
cat_type = CategoricalDtype(categories=category_order, ordered=True)
nypd['laiko_intervalas'] = nypd['laiko_intervalas'].astype(cat_type)

# Suskaičiuojame, kiek kartų pasikartojo kiekvienas laiko intervalas
interval_counts = nypd['laiko_intervalas'].value_counts().sort_index()

# Duomenų paruošimas grafikui
labels = interval_counts.index
counts = interval_counts.values

# Braižome bar diagramą
plt.figure(figsize=(10, 6))
plt.bar(labels, counts, color='lightgreen')
plt.xlabel('Laiko intervalai')
plt.ylabel('Pasikartojimų skaičius')
plt.title('Susišaudymų pasikartojimai pagal paros laiką')

# Pridėti etiketes su pasikartojimų skaičiumi prie stulpelių
for i in range(len(counts)):
    plt.text(i, counts[i] + 5, str(counts[i]), ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# pridėkite naujus stulpelius - kaina, rida kaip skaičiai (integer) 
# su apply() metodu, pridėkite naują stulpelį amžius, kaip skaičius

import pandas as pd
import numpy as np

# SQL užklausa
sql = """select gamintojas, rida, price, variklis, pagaminimo_data from autopliuslt"""
df = pd.read_sql_query(sql, con=DB)

# Ištraukiame tik metus, pirmus keturis stulpelio skaicius ir verciame juos i int
df['metai'] = df['pagaminimo_data'].str[:4].astype(int)  
# Funkcija kainai ir ridai konvertuoti į sveikus skaičius
def convert_price(price):
    if pd.notnull(price):  # Tikriname, ar vertė nėra NaN
        try:
            return int(str(price).replace(',', '').replace(' ', ''))  # Pašaliname tarpus ir kablelius
        except ValueError:
            return np.nan  # Jei konversija nepavyko, grąžiname np.nan
    return np.nan  # Jei vertė netinkama, priskiriame np.nan

def convert_rida(rida):
    if pd.notnull(rida) and 'Nenurodyta' not in str(rida):  # Tikriname, ar vertė nėra NaN ir nėra 'Nenurodyta'
        try:
            return int(str(rida).replace('km', '').replace(' ', ''))  # Pašaliname 'km' ir tarpus
        except ValueError:
            return np.nan  # Jei konversija nepavyko, grąžiname np.nan
    return np.nan  # Jei vertė netinkama, priskiriame np.nan
        
# Priskiriame konvertuotas kainas ir ridą
df['price'] = df['price'].apply(convert_price)
df['rida'] = df['rida'].apply(convert_rida)

# Funkcija amžiui apskaičiuoti
def calculate_age(year):
    if pd.notnull(year):  # Tikriname, ar metai yra teisingi
        current_year = 2024
        return current_year - year
    return np.nan  # Jei metai nėra pateikti, priskiriame np.nan

# Pridedame stulpelį 'amzius' naudodami apply() ir calculate_age funkciją
df['amzius'] = df['metai'].apply(calculate_age)

# Išvedame pirmąsias eilutes, kad patikrintume
df.head()
```
