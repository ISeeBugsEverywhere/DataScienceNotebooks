```python
#duotas failas fpc.csv.
#parašyti funkciją, kuri nuskaitytų failą, ir grąžintų visados pirmą stulpelį ir
# stulpelį, 
# gautą kaip rezultata, kuriuos jūs nurodote kaip funkcijos argumentą (Siuo atveju skirtumo sk.). HINT -
#pasiskaitykite apie eval() funkciją, ir pritaikykite ją čia.
#https://www.programiz.com/python-programming/methods/built-in/eval
#paskutinis stulpelis taip pat turi buti isvestas. KITAS VARIANTAS EVAL PANAUDOJIMO. DESTYTOJO VARIANTAs
import csv

# Funkcija, kuri nuskaitys failą ir grąžins pirmą stulpelį, rezultatų stulpelį pagal eval ir paskutinį stulpelį
def retCols(fn, maths):
    with open(fn, mode='r', encoding='utf-8') as file:
        reader = csv.reader(file)
        header = next(reader)  # Nuskaityti antraštę

        first_col = []
        result_col = []
        last_col = []
        
        for row in reader:
            row = [float(item) for item in row]  # Konvertuoti visus elementus į float
            first_col.append(round(row[0], 2))
            last_col.append(round(row[-1], 2))
            result = eval(maths, {}, {f'c{i+1}': row[i] for i in range(len(row))})
            result_col.append(round(result, 2))
    
    return first_col, result_col, last_col

# Pavyzdinis CSV failo kelias
fn = '../../DATA/fpc.csv'

# Pavyzdinė matematinė išraiška
maths = 'round(c4**2 + c5**2, 6)'

# Kvietimas funkcijai su matematiniais stulpelių pavadinimais
first_col, result_col, last_col = retCols(fn, maths)

# Išspausdiname rezultatus stulpeliais
print(f"{'Pirmas st.':^15}{'Skirtumas':^15}{'Paskutinis st':^15}")
print('-' * 45)
for c1, res, c2 in zip(first_col, result_col, last_col):
    print(f"{c1:^15}{res:^15}{c2:^15}")
```
