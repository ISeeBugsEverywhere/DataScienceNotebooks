```python
# parašyti funkciją, kuri suskaičiuotų pateiktų skaičių aritmetinį vidurkį.
# Skaičiai pateikiami per *args
# skaičiai gali būti kaip tekstai: '1', '2', '3.45280'
# jei bent vieno skaičiaus nesugebėjo konvertuoti: turi grąžinti tekstą '0f'
# vidurkį pateikti visada 4rių ženklų tikslumu
def calculate_average(*args):# * -  neapibreztas skaicius argumentu. Kiek ivesim, tiek ir bus
    numbers = []#kuriamas naujas tuscias sarasas
    for arg in args:
        try:
            number = float(arg)# tikriname ar skaiciai
            numbers.append(number) # surenkame skaicius
        except ValueError:
            return '0f'
    
    if len(numbers) == 0:
        return '0f'# tikriname ar sarase kazkas yra
    
    average = sum(numbers) / len(numbers) #skaiciuojame vidurki
    return f"{average:.4f}"

# Testavimas
print(calculate_average('10', '20', '34.5280657'))
```

```python
# Prisikelkite į darbinį aplanką įvairių failų : paveikslėlių,. csv, tekstinių, docx, pdf failų.
# Paklauskite vartotojo aplanko pavadinimo
# Sukurkite jį, jame sukurkite kitus atlankus, kurių pavadinimai būtų 
# 'Tekstiniai failai', 'Paveikslėliai', 'Wordai', 'pdfai'
# ir tada surūšiuokite savo susirinktus failus į naujojo aplanko atitinkamus subfolderius.
import os
import shutil

def create_and_sort_folders(base_path):
    # Paklauskite vartotojo aplanko pavadinimo
    main_folder = input("Įveskite pagrindinio aplanko pavadinimą: ")
    
    # Pilnas kelias iki pagrindinio aplanko
    main_folder_path = os.path.join(base_path, main_folder)

    # Sukurkite pagrindinį aplanką
    if not os.path.exists(main_folder_path):
        os.mkdir(main_folder_path)

    # Poaplankių pavadinimai
    subfolders = ["Tekstiniai failai", "Paveikslėliai", "Wordai", "pdfai"]

    # Sukuriame poaplankius
    for subfolder in subfolders:
        path = os.path.join(main_folder_path, subfolder)
        if not os.path.exists(path):
            os.mkdir(path)

    # Failų tipų grupavimas
    file_groups = {
        "Tekstiniai failai": ["*.txt"],
        "Paveikslėliai": ["*.jpg", "*.jpeg", "*.png", "*.gif", "*.bmp"],
        "Wordai": ["*.doc", "*.docx"],
        "pdfai": ["*.pdf"]
    }

    # Gauti failus iš nurodyto base_path aplanko
    files = [f for f in os.listdir(base_path) if os.path.isfile(os.path.join(base_path, f))]

    # Perkeliame failus į atitinkamus poaplankius
    for file in files:
        for subfolder, extensions in file_groups.items():
            if any(file.lower().endswith(ext.replace('*', '')) for ext in extensions):
                shutil.move(os.path.join(base_path, file), os.path.join(main_folder_path, subfolder, file))
                break

if __name__ == "__main__":
    base_path = r"C:\Users\eleta\OneDrive\Desktop\Mokslai\DataScienceNotebooks\Studentai\EleTame"
    create_and_sort_folders(base_path)
```

```python
# Jūs turite aplanką DATA su įvairiais failais. parašykite Python kodą, kuris padarytų šio aplanko kopiją,
# prie kiekvieno failo pavadinimo pradžios pridėdamas po tekstą "09p_"
import os
import shutil
def copy_and_rename_folder(src_folder, dst_folder, prefix):
    # Sukuriameas paskirties aplanką, jei jo nėra
    if not os.path.exists(dst_folder):
        os.makedirs(dst_folder)

    # Pereiname per visus failus ir aplankus šaltinio aplanke
    for item in os.listdir(src_folder):
        src_path = os.path.join(src_folder, item)
        dst_path = os.path.join(dst_folder, prefix + item)

        if os.path.isdir(src_path):
            # Kopijuojame aplankus
            copy_and_rename_folder(src_path, dst_path, prefix)
        else:
            # Kopijuojame ir pervadiname failus
            shutil.copy2(src_path, dst_path)

if __name__ == "__main__":
    #tiksliai nurodau, kur yra DATA byla ir kur turiu irasyti nauja faila
    src_folder = r"C:\Users\eleta\OneDrive\Desktop\Mokslai\DataScienceNotebooks\DATA"
    dst_folder = r"C:\Users\eleta\OneDrive\Desktop\Mokslai\DataScienceNotebooks\DATA_copy"
    prefix = "09p_"
    
    copy_and_rename_folder(src_folder, dst_folder, prefix)
```

```python
# Parašykite kodą, kuris ištrintų aplanką, kurio pavadinimą (arba kelią iki jo) 
# įveda vartotojas.
# Kartu parašykite, kiek failų ištrynė, del to, kad buvo istrintas aplankas
import os
import shutil

def delete_folder(folder_path):
    # Tikriname, ar aplankas egzistuoja
    if not os.path.exists(folder_path):
        print(f"Aplanko '{folder_path}' nėra.")
        return
    
    file_count = 0
    
    # Naudojame os.walk() su topdown=False, kad galėtume saugiai ištrinti failus ir aplankus iš apačios į viršų.
    for root, dirs, files in os.walk(folder_path, topdown=False):
        # Pridedame failų skaičių dabartiniame kataloge prie file_count
        file_count = len(files) +1
        
        for name in files:
            os.remove(os.path.join(root, name))
        for name in dirs:
            os.rmdir(os.path.join(root, name))
    
    # Galiausiai ištriname pagrindinį aplanką
    shutil.rmtree(folder_path)
    
    print(f"Aplankas '{folder_path}' sėkmingai ištrintas.")
    if file_count == 0:
        print(f"Aplanke nebuvo failų.")
    else:
        print(f"Ištrinta failų: {file_count}")
if __name__ == "__main__":
    folder_name = input("Įveskite aplanko pavadinimą: ")
    delete_folder(folder_name)
```

```python
 # programa turi sukurti aplanką, kurio pavadinimą įvestų pats vartotojas.
 #Patikrina ar tokiu pavadinimu aplankas egzistuoja 

import os

# Paprašyti vartotojo įvesti aplanko pavadinimą
name = input("Įveskite aplanko pavadinimą: ")

# Patikrinti, ar aplankas jau egzistuoja
if not os.path.exists(name):
    # Sukurti aplanką
    os.makedirs(name)
    print(f"Aplankas '{name}' sukurtas sėkmingai.")
else:
    print(f"Aplankas '{name}' jau egzistuoja.")
```
